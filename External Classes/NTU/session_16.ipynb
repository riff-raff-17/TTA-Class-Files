```python
from final_helpers import AP_centralization_approaching, pick_up, face_find_and_approach

from ugot import ugot
got = ugot.UGOT()
got.initialize("192.168.1.1")

got.load_models(["apriltag_qrcode", "line_recognition", 
                 "face_recognition", "word_recognition"])
```

```python
# 1) AprilTag recognition
print("Starting AprilTag recognition...")
AP_centralization_approaching(got, distance=0.18, gap=20, fwd_spd=5, 
                              strafe_spd=5)
pick_up(got)
print("Finished picking up AprilTag!")
```

```python
def line_follow(mult=0.25, speed=20):
    offset, line_type, _, _ = got.get_single_track_total_info()
    rotation_speed = int(offset * mult)
    got.mecanum_move_xyz(0, speed, rotation_speed)
    return line_type
```

```python
import time

# 1) AprilTag recognition
print("Starting AprilTag recognition...")
AP_centralization_approaching(got, distance=0.11, gap=20, fwd_spd=5, 
                              strafe_spd=5)
pick_up(got)
print("Finished picking up AprilTag!")

# 2) Line follow
print("Starting line following...")
got.set_track_recognition_line(0)
while True:
    line = line_follow()
    if line == 0:
        got.mecanum_stop()
        print("End of line!")
        break

got.mechanical_joint_control(angle1=0, angle2=-30, angle3=-30, duration=500)
time.sleep(1)
got.mechanical_clamp_release()
```
