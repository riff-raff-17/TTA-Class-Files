```python
from ugot import ugot
got = ugot.UGOT()
got.initialize("192.168.1.128")

got.load_models(["face_recognition", "word_recognition"])

got.open_camera()
```

```python
# Add in a new face
got.face_recognition_add_name(name="Bob")
```

```python
# Delete a face
got.face_recognition_delete_name(name="Bob")
```

```python
# Get a list of all faces
got.face_recognition_get_all_names()
```

```python
# Delete all names
name_lst = got.face_recognition_get_all_names()

for name in name_lst:
    got.face_recognition_delete_name(name=name)
```

```python
import cv2
import numpy as np
from IPython.display import display, Image, clear_output

try:
    while True:
        # Grab the latest camera frame as raw bytes
        frame = got.read_camera_data()

        # Decode the bytes into an OpenCV image array
        nparr = np.frombuffer(frame, np.uint8)
        img = cv2.imdecode(nparr, cv2.IMREAD_COLOR)

        # Get face recognition result:
        # [[name, center_x, center_y, height, width, area], ...]
        faces = got.get_face_recognition_total_info()

        # Process each face and draw it on the camera
        if faces:
            for face in faces: 
                name, cx, cy, h, w, area = face
                cx, cy = int(cx), int(cy)
                half_w, half_h = int(w / 2), int(h / 2)

                # Compute bounding box corners
                x1, y1 = cx - half_w, cy - half_h
                x2, y2 = cx + half_w, cy + half_h

                # Draw bounding box and center dot
                cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.circle(img, (cx, cy), 4, (0, 0, 255), -1)

                # Label with name
                label = f"{name}  area: {area}px"
                cv2.putText(
                    img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.7,
                    (0, 255, 0), 2
                )

        # Encode the annotated frame as a JPEG and display it inline in the notebook
        _, jpeg = cv2.imencode(".jpg", img)
        clear_output(wait=True)
        display(Image(data=jpeg.tobytes()))

except KeyboardInterrupt:
    print("Done")
    got.mecanum_stop()
```
