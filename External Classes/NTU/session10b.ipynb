```python
from ugot import ugot
got = ugot.UGOT()
got.initialize("192.168.1.20")
```

```python
import time

def pick_up():
    print("Starting pick up code...")
    got.mechanical_clamp_release()
    got.mechanical_joint_control(angle1=0, angle2=-30, angle3=-60, duration=500)
    time.sleep(1)
    got.mechanical_clamp_close()
    time.sleep(1)
    got.mechanical_joint_control(angle1=0, angle2=90, angle3=0, duration=1000)
    time.sleep(1)
    print("Picked up tag successfully!")

def put_down():
    print("Starting put down code...")
    got.mechanical_joint_control(angle1=0, angle2=-25, angle3=-25, duration=1500)
    time.sleep(1.5)
    got.mechanical_clamp_release()
    time.sleep(0.5)
    got.mechanical_joint_control(angle1=0, angle2=45, angle3=45, duration=500)
    time.sleep(0.5)
    print("Dropped off succesfully!")
```

```python
pick_up()
got.mecanum_move_speed_times(direction=0, speed=30, times=100, unit=1)
put_down()
```
