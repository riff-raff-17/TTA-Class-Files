```python
from ugot import ugot
got = ugot.UGOT()
got.initialize("192.168.88.1")
```

```python
import time

got.mechanical_clamp_release() # opens the clamp
time.sleep(1)
got.mechanical_clamp_close() # closes the clamp
time.sleep(1)
got.mechanical_arms_restory() # resets the arm
time.sleep(1)
got.mechanical_joint_control(angle1=0, angle2=45, 
                             angle3=45, duration=500)
```

```python
got.mechanical_arms_restory() # reset the arm
time.sleep(1)
got.mechanical_clamp_release() # open the clamp
time.sleep(1)
got.mechanical_joint_control(angle1=0, angle2=-45,
                             angle3=-45, duration=500)
time.sleep(1)
got.mechanical_clamp_close() # close the clamp
time.sleep(1)
got.mechanical_joint_control(angle1=0, angle2=45,
                             angle3=45, duration=500)
```
