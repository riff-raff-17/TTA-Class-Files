```python
from ugot import ugot
got = ugot.UGOT()
got.initialize("192.168.88.1")
print("Connected")
```

```python
import time

# UGOT lights
got.show_light_rgb(lights=[0, 1, 2, 3], red=0, green=255, blue=0)

# Print out text on the screen
got.screen_print_text_newline(text="I am a robot", color=1)

# Start robot
got.wheelleg_start_balancing()
time.sleep(2)

while True:
    cmd = input("What should I do?")

    # Highest priority (top to bottom)
    if cmd in {"q", "quit", "exit"}:
        print("Done!")
        break

    elif cmd in {"f", "fwd", "forward", "w"}:
        got.wheelleg_move_speed_times(direction=0, speed=30, times=30, unit=1)
```
