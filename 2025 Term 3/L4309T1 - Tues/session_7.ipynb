```python
import tkinter as tk
from ugot import ugot
got = ugot.UGOT()
got.initialize('192.168.1.198')

def say_hello():
    user = entry.get()
    if user.strip() == "Wenhao":
        label_output.config(text="Wenhao is growing a garden")
    elif user.strip() == "Wenxuan":
        label_output.config(text="Wenxuan is blooking a blooklet")
    elif user.strip() == "Caden":
        label_output.config(text="CCCCCCaden")
    elif user.strip() == "Austin":
        label_output.config(text="AustinAnyAwesome eats McDonald's")
    elif user.strip():
        label_output.config(text=f"Hello, {user}!")
        got.play_audio_tts(data=f"Hello, {user}!", voice_type=0, wait=True)
    else:
        label_output.config(text="Please enter your name")

# Create the main window
root = tk.Tk()
root.title("Hello App")
root.geometry("300x200")

# Create widgets
label_prompt = tk.Label(root, text="Enter your name: ")
entry = tk.Entry(root)
button = tk.Button(root, text="Say Hello", command=say_hello)
label_output = tk.Label(root, text="", fg='white')

# Position widgets on the screen
label_prompt.pack(pady=5)
entry.pack(pady=5)
button.pack(pady=5)
label_output.pack(pady=10)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk

def say_hello(*_):
    name = name_var.get().strip()
    if name:
        output_var.set(f"Hello, {name}!")
        output_label.configure(style="Success.TLabel")
    else:
        output_var.set("Please enter your name.")
        output_label.configure(style="Warn.TLabel")

# --- window ---
root = tk.Tk()
root.title("Hello App")
root.geometry("320x200")

# --- theme + styles ---
style = ttk.Style(root)

# Pick a decent built-in theme if available
for candidate in ("clam", "vista", "xpnative", "alt", "default"):
    if candidate in style.theme_names():
        style.theme_use(candidate)
        break

# Base widget tweaks
style.configure("TButton", padding=(10, 6))
style.configure("TEntry", padding=6)

# A slightly bolder title label
style.configure("Title.TLabel", font=("TkDefaultFont", 12, "bold"))

# Output styles
style.configure("Success.TLabel", foreground="#166534")  # green-700-ish
style.configure("Warn.TLabel",    foreground="#7c2d12")  # amber-900-ish

# --- layout container ---
container = ttk.Frame(root, padding=12)
container.pack(fill="both", expand=True)

# --- widgets ---
ttk.Label(container, text="Hello App", style="Title.TLabel").pack(anchor="w", pady=(0, 6))

name_var = tk.StringVar()
output_var = tk.StringVar(value="")  # weâ€™ll style this label dynamically

prompt = ttk.Label(container, text="Enter your name:")
entry = ttk.Entry(container, textvariable=name_var, width=28)
button = ttk.Button(container, text="Say Hello", command=say_hello)
output_label = ttk.Label(container, textvariable=output_var, style="Warn.TLabel")

# --- layout ---
prompt.pack(anchor="w")
entry.pack(fill="x", pady=6)
button.pack(pady=(0, 8))
output_label.pack()

# --- UX niceties ---
entry.focus_set()
entry.bind("<Return>", say_hello)    # press Enter to submit
root.bind("<Escape>", lambda e: root.destroy())  # quick exit

root.mainloop()
```
