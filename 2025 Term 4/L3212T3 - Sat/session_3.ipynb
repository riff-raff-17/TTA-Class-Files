```python
import math

def circle(radius):
    # circumference of a circle = 2 * pi * r OR pi * d
    circumference = 2 * math.pi * radius
    n = 30 # how precise our circle is
    length = circumference / n
    polygon(n, length)

def polygon(n, length):
    angle = 360 / n
    for _ in range(n):
        forward(length)
        left(angle)

make_turtle(height=300, width=300,delay=0.02)
set_color("purple")
circle(60)
```

```python
# multicircle
# range * left must equal 360 in order to make a full circle
n = 30
make_turtle(delay=0.001)
for i in range(n):
    circle(30)
    left(360 / n)
```

```python
def polyline(n, length, angle):
    for _ in range(n):
        forward(length)
        left(angle)

def arc(radius, angle):
    arc_length = 2 * math.pi * radius * angle / 360
    n = 30 # precision of the circle
    length = arc_length / n
    step_angle = angle / n
    polyline(n, length, step_angle)

def petal(radius, angle):
    for i in range(2):
        arc(radius, angle)
        left(180 - angle)

make_turtle(delay=0.02)
petal(30, 90)
```

```python
def flower(n, radius, angle):
    for i in range(n):
        petal(radius, angle)
        left(360 / n)

make_turtle(delay=0.002)
flower(23, 50, 90)
```
