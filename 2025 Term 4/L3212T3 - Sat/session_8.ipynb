```python
from getpass import getpass

users = {"john dough" : "ineedfreerobux123",
         "admin" : "password1",
         "qwertyuiop": "13579"}

def login(users, max_attempts=3):
    username = input("Username: ").strip()
    if username not in users:
        print("User not found!")
        return False
    
    for attempt in range(max_attempts):
        pwd = getpass("Password: ").strip()
        if users[username] == pwd:
            print(f"Login successful! Welcome, {username}")
            return True
        else:
            left = max_attempts - attempt - 1
            print(f"Incorrect password. {left} attempts left")

    print("Account locked. Self destructing...")
    return False

login(users)
```

```python
from ugot import ugot
import time, random
from IPython.display import clear_output

got = ugot.UGOT()
got.initialize("192.168.1.217")

try:
    while True:
        print(got.read_distance_data(21))
        time.sleep(0.1)
        clear_output(wait=True)
except KeyboardInterrupt:
    got.mecanum_stop()
    print("Done")
```
