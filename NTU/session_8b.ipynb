```python
from ugot import ugot
got = ugot.UGOT()
got.initialize("192.168.1.126")
print("Done!")
```

```python
import time

# Open gripper
got.mechanical_clamp_release()
time.sleep(1)

# Close gripper
got.mechanical_clamp_close()

# Move the arm
got.mechanical_joint_control(angle1=0, angle2=0, angle3=0, duration=500)
time.sleep(0.5)

# Single joint control
got.mechanical_single_joint_control(joint=1, angle=0, duration=500)
time.sleep(0.5)
```

```python
# Picking up
got.mechanical_joint_control(angle1=0, angle2=-45, angle3=-45, duration=500)
got.mechanical_clamp_release()
time.sleep(2) # How long to pickup object

got.mechanical_clamp_close()
time.sleep(1)

# Underhand throw
got.mechanical_joint_control(angle1=0, angle2=45, angle3=45, duration=150)
got.mechanical_clamp_release()
```
