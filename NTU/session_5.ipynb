```python
prize = ["gold", "silver", "bronze"]

# Wegihts
weights = [1, 5, 10]
weights = [0.05, 0.2, 0.75]

random.choices(prize, weights=weights, k=5)
```

```python
# List of possible moves
moves = ["rock", "paper", "scissors"]
weights = [1, 1, 1]

# computer move
computer = random.choices(moves, weights=weights, k=1)[0]

player = input("Choose rock, paper, or scissors:").strip().lower()

print("You chose:", player)
print("Computer chose:", computer)

if (player == 'paper' and computer == 'rock') or \
    (player == 'scissors' and computer == 'paper') or \
    (player == 'rock' and computer == 'scissors'):

    print("You won!")

elif player == computer:
    print("It is a draw ._.")

else:
    print("You lose :(")
```

```python
from ugot import ugot
from IPython.display import clear_output
got = ugot.UGOT()
got.initialize("192.168.1.1")
```

```python
import time

got.wheelleg_start_balancing()
time.sleep(2)

try:
    while True:
        forward = random.randint(0, 1)
        dist = random.randint(0, 100)
        got.wheelleg_move_speed_times(direction=forward, speed=20, times=dist, unit=1)

        turn = random.randint(2, 3)
        angle = random.randint(45, 180)
        got.wheelleg_turn_speed_times(turn=turn, speed=45, times=angle, unit=2)

except KeyboardInterrupt:
    got.wheelleg_stop_balancing()
    print("Done!")
```
