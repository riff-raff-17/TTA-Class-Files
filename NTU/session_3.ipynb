```python
from ugot import ugot
got = ugot.UGOT()
got.initialize("192.168.1.125")
print("Connected")
```

```python
import time

got.wheelleg_start_balancing()
time.sleep(2)

command = input("What is your command?")

if command == 'forward':
    # 0 is Forwards, 1 is Backwards
    got.wheelleg_move_speed_times(direction=0, speed=30, times=20, unit=1)
    print("Moving forward")
elif command == "backward":
    # 2 is Left, 3 is Right
    got.wheelleg_move_speed_times(direction=1, speed=30, times=20, unit=1)
    print("Moving backward")
elif command == "left":
    # 2 is Left, 3 is Right
    got.wheelleg_turn_speed_times(turn=2, speed=45, times=90, unit=2)
elif command == "right":
    # 2 is Left, 3 is Right
    got.wheelleg_turn_speed_times(turn=3, speed=45, times=90, unit=2)
else:
    print("Command not recognized")
```
