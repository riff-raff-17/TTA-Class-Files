```python
# connect and initialize robot
from ugot import ugot
got = ugot.UGOT()
got.initialize("192.168.1.1")
print("Connected!")
```

```python
import time

# move
got.wheelleg_start_balancing()
time.sleep(1)

# 1 is high, 2 is medium, 3 is low
got.wheelleg_set_chassis_height(height=3)

# move forward/back
# direction: 0 is forwards, 1 is back
got.wheelleg_move_speed_times(direction=0, speed=40, times=100, unit=1)

# turn left/right
# turn: 2 is left, 3 is right
got.wheelleg_turn_speed_times(turn=2, speed=45, times=90, unit=2)
```
