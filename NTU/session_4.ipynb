```python
from ugot import ugot
got = ugot.UGOT()
got.initialize('192.168.1.125')
print("Connected!")
```

```python
import time

# Start the robot and keep it balancing upright
got.wheelleg_start_balancing()
time.sleep(2) # give it two seconds to stand up

# 0 = forwards, 1 = backwards
got.wheelleg_move_speed_times(direction=0, speed=30, times=50, unit=1)

# 2 = left, 3 = right
got.wheelleg_turn_speed_times(turn=2, speed=45, times=90, unit=2)
```

```python
from IPython.display import clear_output
import time

try:
    while True:
        distance = got.read_distance_data(61)
        print("Distance:", got.read_distance_data(61), "cm")

        if distance <= 15:
            got.screen_display_background(color=3)
            got.wheelleg_stop_balancing()
            got.wheelleg_turn_speed_times(turn=2, speed=45, times=90, unit=2)
        else:
            got.screen_display_background(color=6)
            got.wheelleg_move_speed(direction=0, speed=30)

        clear_output(wait=True)
except KeyboardInterrupt:
    got.wheelleg_stop_balancing()
    print("Done!")
```
