```python
from ugot import ugot
from IPython.display import clear_output
got = ugot.UGOT()
got.initialize("192.168.1.1")

print("Connected!")
```

```python
# Forwards/backwards
got.mecanum_move_speed_times(direction=0, speed=30, times=50, unit=1)

# Left/right
got.mecanum_turn_speed_times(turn=2, speed=45, times=90, unit=2)

# Strafing movement
got.mecanum_translate_speed_times(angle=45, speed=30, times=50, unit=1)
```
