```python
from ugot import ugot
got = ugot.UGOT()
got.initialize("192.168.1.222")
print("Connected!")
```

```python
import whisper
import sounddevice as sd
import soundfile as sf

# 5 models: tiny/base/small/medium/large
model = whisper.load_model("base")

sample_rate = 8000
sec = 3

audio = sd.rec(int(sec * sample_rate), samplerate=sample_rate, channels=1, dtype="float32")
print("Listening!")
sd.wait()
sf.write("temp.wav", audio, sample_rate)

result = model.transcribe("temp.wav")
phrase = result["text"].strip() # strip() removes whitespace
print(f"You said: {phrase}")

command, *rest = phrase.split()
action = " ".join(rest)

if command == "Say":
    print(f"Saying \"{action}\"")
    got.play_audio_tts(data=action, voice_type=0, wait=True)
if command == "Go":
    print(f"Going \"{action}\"")
if command == "Do":
    print(f"Doing \"{action}\"")
```

```python
direction, content = got.start_audio_asr_doa(duration=2)

print(f"Direction: {direction}\nContent: {content}")

if content:
    got.play_audio_tts(data=content, voice_type=0, wait=True)

if direction == "Front":
    got.balance_turn_speed_times(turn=2, speed=45, times=90, unit=2)
elif direction == "Behind":
    got.balance_turn_speed_times(turn=3, speed=45, times=90, unit=2)
elif direction == "Left":
    got.balance_move_speed_times(direction=0, speed=30, times=50, unit=1)
elif direction == "Right":
    got.balance_move_speed_times(direction=1, speed=30, times=50, unit=1)
```

### Sentiment analysis

__Polarity__ is how positive or negative the statement is. -1.0 is max negative, and 1.0 is max positive.

__Subjectivity__ is how factually true a statement is. 0.0 is a fact, and 1.0 is an opinion.

```python
text = "Seth is happy gamer"

blob = TextBlob(text)

blob.sentiment
```

```python
print("I will detect your feelings and facts")
print("Tell me something:")

# Get user input
user_input = input("Input your statement >")

blob = TextBlob(user_input)
sentiment = blob.sentiment.polarity # positive or negative statement
truthfulness = blob.sentiment.subjectivity # fact or opinion

if sentiment > 0:
    print("Happy statement!")
elif sentiment < 0:
    print("Sad statement :(")
else:
    print("Neutral statement ._.")
```
