```python
"""Hangman"""

secret = "robot"

lives = 6
guessed_letters = ""
display = ["_"] * len(secret)  # _ _ _ _ _

print("=== Hangman ===")
print("I'm thinking of a word with", len(secret), "letters")

try:
    while True:
        print("\nWord:", " ".join(display))
        print("Lives left:", lives)
        print("Guessed so far:", " ".join(guessed_letters))

        guess = (
            input("Guess a single letter: ").lower().strip()
        )  # lower() = lowercase, strip() removes whitespace

        if guess == "quit":
            print("Goodbye!")
            break
        if len(guess) != 1 or not guess.isalpha():  # isalpha = alphabetical
            print("Enter one letter!")
            continue
        if guess in guessed_letters:
            print("Already guessed!")
            continue

        guessed_letters += guess

        if guess in secret:
            for i in range(len(secret)):
                if secret[i] == guess:
                    display[i] = guess
            print("u liv")
        else:
            lives -= 1  # "-= 1" means to subtract one from lives
            print("not noice :(")

        if "_" not in display:
            print("\nYou win! The word was:", secret)
        if lives == 0:
            print("\nOut of lives! The word was:", secret)
            break

except KeyboardInterrupt:
    print("Done")
```
