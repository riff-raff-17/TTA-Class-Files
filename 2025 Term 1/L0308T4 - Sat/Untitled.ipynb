```python
#Import UGOT packages
from ugot import ugot

#Rename functions
got = ugot.UGOT()

#Initialize UGOT
got.initialize('192.168.1.235')
```

```python
got.transform_adaption_control(False)

#Normal position
#Arm (1: left front arm; 2: left rear arm; 3: right rear arm; 4: right front arm)
def normal():
    got.transform_arm_control(joint=1, position=-40, time=500) #NEGATIVE IS DOWN
    got.transform_arm_control(joint=4, position=40, time=500) #POSITIVE IS DOWN
    got.transform_arm_control(joint=2, position=40, time=500) #POSITIVE IS DOWN
    got.transform_arm_control(joint=3, position=-40, time=500) #NEGATIVE IS DOWN

def front_up():
    got.transform_arm_control(joint=1, position=40, time=500) #NEGATIVE IS DOWN
    got.transform_arm_control(joint=4, position=-40, time=500) #POSITIVE IS DOWN
    got.transform_arm_control(joint=2, position=40, time=500) #POSITIVE IS DOWN
    got.transform_arm_control(joint=3, position=-40, time=500) #NEGATIVE IS DOWN

def back_up():
    got.transform_arm_control(joint=1, position=-40, time=500) #NEGATIVE IS DOWN
    got.transform_arm_control(joint=4, position=40, time=500) #POSITIVE IS DOWN
    got.transform_arm_control(joint=2, position=-40, time=500) #POSITIVE IS DOWN
    got.transform_arm_control(joint=3, position=40, time=500) #NEGATIVE IS DOWN

def climb_obstacle():
    got.transform_adaption_control(False)

    front_up()
    got.transform_move_speed_times(direction=0, speed=30, times=20, unit=1)
    normal()
    got.transform_move_speed_times(direction=0, speed=30, times=20, unit=1)
    back_up()
    got.transform_move_speed_times(direction=0, speed=30, times=20, unit=1)
    normal()
    got.transform_move_speed_times(direction=0, speed=30, times=20, unit=1)
    got.transform_move_speed_times(direction=0, speed=30, times=20, unit=1)

def under_bridge():
    got.transform_arm_control(joint=1, position=-25, time=500) #NEGATIVE IS DOWN
    got.transform_arm_control(joint=4, position=25, time=500) #POSITIVE IS DOWN
    got.transform_arm_control(joint=2, position=25, time=500) #POSITIVE IS DOWN
    got.transform_arm_control(joint=3, position=-25, time=500) #NEGATIVE IS DOWN
    
    got.transform_move_speed_times(direction=0, speed=30, times=100, unit=1) 

    got.transform_arm_control(joint=1, position=-90, time=500) #NEGATIVE IS DOWN
    got.transform_arm_control(joint=4, position=90, time=500) #POSITIVE IS DOWN
    got.transform_arm_control(joint=2, position=90, time=500) #POSITIVE IS DOWN
    got.transform_arm_control(joint=3, position=-90, time=500) #NEGATIVE IS DOWN

climb_obstacle()
under_bridge()
```

```python
import pygame

# Initialize pygame
pygame.init()

# Constants
WIDTH, HEIGHT = 800, 600
BALL_RADIUS = 10
PADDLE_WIDTH, PADDLE_HEIGHT = 10, 100
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
FPS = 60
BALL_SPEED_X, BALL_SPEED_Y = 4, 4
PADDLE_SPEED = 6
WINNING_SCORE = 10
SPEED_INCREMENT = 1
MAX_BALL_SPEED = 10

# Create window
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Pong")

# Font for score
font = pygame.font.Font(None, 36)

# Paddle positions
left_paddle = pygame.Rect(30, HEIGHT//2 - PADDLE_HEIGHT//2, PADDLE_WIDTH, PADDLE_HEIGHT)
right_paddle = pygame.Rect(WIDTH - 40, HEIGHT//2 - PADDLE_HEIGHT//2, PADDLE_WIDTH, PADDLE_HEIGHT)
ball = pygame.Rect(WIDTH//2 - BALL_RADIUS, HEIGHT//2 - BALL_RADIUS, BALL_RADIUS*2, BALL_RADIUS*2)

# Ball velocity
ball_dx, ball_dy = BALL_SPEED_X, BALL_SPEED_Y

# Scores
left_score = 0
right_score = 0

# Game loop
running = True
clock = pygame.time.Clock()
while running:
    screen.fill(BLACK)
    pygame.draw.rect(screen, WHITE, left_paddle)
    pygame.draw.rect(screen, WHITE, right_paddle)
    pygame.draw.ellipse(screen, WHITE, ball)
    pygame.draw.aaline(screen, WHITE, (WIDTH//2, 0), (WIDTH//2, HEIGHT))
    
    # Display scores
    left_text = font.render(str(left_score), True, WHITE)
    right_text = font.render(str(right_score), True, WHITE)
    screen.blit(left_text, (WIDTH//4, 20))
    screen.blit(right_text, (3 * WIDTH//4, 20))
    
    # Move ball
    ball.x += ball_dx
    ball.y += ball_dy
    
    # Ball collision with top and bottom walls
    if ball.top <= 0 or ball.bottom >= HEIGHT:
        ball_dy *= -1
    
    # Ball collision with paddles
    if ball.colliderect(left_paddle) or ball.colliderect(right_paddle):
        ball_dx *= -1
        if abs(ball_dx) < MAX_BALL_SPEED:
            ball_dx += SPEED_INCREMENT if ball_dx > 0 else -SPEED_INCREMENT
        if abs(ball_dy) < MAX_BALL_SPEED:
            ball_dy += SPEED_INCREMENT if ball_dy > 0 else -SPEED_INCREMENT
    
    # Ball out of bounds (score tracking and reset position)
    if ball.left <= 0:
        right_score += 1
        ball.x, ball.y = WIDTH//2 - BALL_RADIUS, HEIGHT//2 - BALL_RADIUS
        ball_dx, ball_dy = BALL_SPEED_X, BALL_SPEED_Y
        ball_dx *= -1
    if ball.right >= WIDTH:
        left_score += 1
        ball.x, ball.y = WIDTH//2 - BALL_RADIUS, HEIGHT//2 - BALL_RADIUS
        ball_dx, ball_dy = BALL_SPEED_X, BALL_SPEED_Y
        ball_dx *= -1
    
    # Check for winner
    if left_score >= WINNING_SCORE or right_score >= WINNING_SCORE:
        winner_text = font.render("Player Left Wins!" if left_score >= WINNING_SCORE else "Player Right Wins!", True, WHITE)
        screen.blit(winner_text, (WIDTH//2 - 100, HEIGHT//2))
        pygame.display.flip()
        pygame.time.delay(3000)
        running = False
    
    # Paddle movement
    keys = pygame.key.get_pressed()
    if keys[pygame.K_w] and left_paddle.top > 0:
        left_paddle.y -= PADDLE_SPEED
    if keys[pygame.K_s] and left_paddle.bottom < HEIGHT:
        left_paddle.y += PADDLE_SPEED
    if keys[pygame.K_UP] and right_paddle.top > 0:
        right_paddle.y -= PADDLE_SPEED
    if keys[pygame.K_DOWN] and right_paddle.bottom < HEIGHT:
        right_paddle.y += PADDLE_SPEED
    
    # Event handling
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    
    pygame.display.flip()
    clock.tick(FPS)

pygame.quit()
```
