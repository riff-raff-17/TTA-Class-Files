```python
import time

def live_clock():
    print("Live Clock (Press Ctrl+C to exit):")
    try:
        while True:
            current_time = time.strftime("%H:%M:%S")
            print(f"\r{current_time}", end="", flush=True)
            time.sleep(1)
    except KeyboardInterrupt:
        print("\nClock stopped.")

live_clock()
```

```python
import time
import sys

def loading_screen():
    total_steps = 50  
    animation = ["|", "/", "-", "\\"]
    
    print("\nLoading...", end="", flush=True)

    for i in range(total_steps + 1): 
        time.sleep(0.1) 
        progress = i / total_steps
        percent = f"{int(progress * 100)}%".rjust(4) 
        spinner = animation[i % len(animation)]
        sys.stdout.write(f"\rLoading {spinner} {percent}")
        sys.stdout.flush()

    print("\nLoading complete!")

loading_screen()
```
