```python
from jupyturtle import make_turtle, forward, back, left, right, draw, set_color
```

```python
make_turtle()

#square
for i in range(4):
    forward(50)
    left(90)
```

```python
make_turtle()

#triangle
for i in range(3):
    forward(50)
    left(120)
```

```python
make_turtle(width=300, height=200)

#pentagon
for i in range(5):
    forward(50)
    left(72)
```

```python
def polygon(sides, length):
    angle = 360 / sides
    for i in range(sides):
        forward(length)
        left(angle)

make_turtle(animate=False,height=400)
polygon(200, 4)
draw()
```

```python
import math

def circle(radius):
    circumference = 2 * math.pi * radius
    #circle precision
    n = 30
    length = circumference / n
    polygon(n, length)

make_turtle(delay=0.02)
circle(30)
```

```python
make_turtle(delay = 0.001)

for i in range(30):
    circle(30)
    left(12)
```

```python
colors = 'red orange gold green blue rebeccapurple'.split()

def flower_circle(n, radius):
    for i in range(n):
        set_color(colors[i % len(colors)])
        circle(radius)
        left(360 / n)
make_turtle(delay=0.001)
flower_circle(5, 30)
```
