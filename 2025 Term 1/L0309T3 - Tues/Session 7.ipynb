```python
import math

def polyline(n, length, angle):
    for i in range(n):
        forward(length)
        left(angle)

def arc(radius, angle):
    arc_length = 2 * math.pi * radius * (angle / 360)
    #circle precision
    n = 30
    length = arc_length / n
    step_angle = angle / n
    polyline(n, length, step_angle)

def circle(radius):
    arc(radius, 360)

make_turtle(delay = 0.01, width=200, height=200)

def multicircle(n, radius):
    for i in range(n):
        circle(radius)
        right(360 / n)

multicircle(30, 30)
```

```python
import random

colors = 'black aqua purple blueviolet blue'.split()

def petal(radius, angle):
    for _ in range(2):
        arc(radius, angle)
        left(180 - angle)

def flower(n, radius, angle):
    for _ in range(n):
        set_color(random.choice(colors))
        petal(radius, angle)
        left(360 / n)

make_turtle(delay=0.002)
flower(10, 50, 90)
```
