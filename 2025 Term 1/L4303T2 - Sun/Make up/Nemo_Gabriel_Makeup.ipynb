```python
# Base game

import pygame
import random

# Initialize pygame
pygame.init()

# Constants
WIDTH, HEIGHT = 600, 600
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)
PLAYER_SIZE = 30
BULLET_SIZE = 10
PLAYER_SPEED = 5
BULLET_SPEED = 3

# Setup screen
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Dodge the Bullets!")

# Player setup
player = pygame.Rect(WIDTH//2, HEIGHT//2, PLAYER_SIZE, PLAYER_SIZE)

# Bullet list
bullets = []
clock = pygame.time.Clock()
running = True

def spawn_bullet():
    side = random.choice(['top', 'bottom', 'left', 'right'])
    if side == 'top':
        x, y = random.randint(0, WIDTH), 0
        dx, dy = 0, BULLET_SPEED
    elif side == 'bottom':
        x, y = random.randint(0, WIDTH), HEIGHT
        dx, dy = 0, -BULLET_SPEED
    elif side == 'left':
        x, y = 0, random.randint(0, HEIGHT)
        dx, dy = BULLET_SPEED, 0
    else:  # right
        x, y = WIDTH, random.randint(0, HEIGHT)
        dx, dy = -BULLET_SPEED, 0
    bullets.append({'rect': pygame.Rect(x, y, BULLET_SIZE, BULLET_SIZE), 'dx': dx, 'dy': dy})

# Game loop
while running:
    screen.fill(WHITE)
    
    # Handle events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Move player
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT] and player.x > 0:
        player.x -= PLAYER_SPEED
    if keys[pygame.K_RIGHT] and player.x < WIDTH - PLAYER_SIZE:
        player.x += PLAYER_SPEED
    if keys[pygame.K_UP] and player.y > 0:
        player.y -= PLAYER_SPEED
    if keys[pygame.K_DOWN] and player.y < HEIGHT - PLAYER_SIZE:
        player.y += PLAYER_SPEED
    
    # Spawn bullets over time
    if random.random() < 0.02:
        spawn_bullet()
    
    # Move bullets
    for bullet in bullets[:]:
        bullet['rect'].x += bullet['dx']
        bullet['rect'].y += bullet['dy']
        
        # Check collision
        if player.colliderect(bullet['rect']):
            running = False
        
        # Remove bullets out of bounds
        if (bullet['rect'].x < -BULLET_SIZE or bullet['rect'].x > WIDTH or 
            bullet['rect'].y < -BULLET_SIZE or bullet['rect'].y > HEIGHT):
            bullets.remove(bullet)
    
    # Draw player
    pygame.draw.rect(screen, BLACK, player)
    
    # Draw bullets
    for bullet in bullets:
        pygame.draw.rect(screen, RED, bullet['rect'])
    
    # Refresh screen
    pygame.display.flip()
    clock.tick(60)

pygame.quit()
exit()
```

```python
#Add x2 zone

import pygame
import random
import time

# Initialize pygame
pygame.init()

# Constants
WIDTH, HEIGHT = 600, 600
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)
BLUE = (0, 0, 255, 100)  # Transparent blue for double score zone
PLAYER_SIZE = 30
BULLET_SIZE = 10
PLAYER_SPEED = 5
BULLET_SPEED = 3
BULLET_INCREMENT = 0.1  # Speed increase per interval
SPAWN_CHANCE_INCREMENT = 0.005  # Increase bullet spawn chance over time
DIFFICULTY_INTERVAL = 3000  # Increase difficulty every 3 seconds
BULLET_SPAWN_CHANCE = 0.02  # Initial spawn chance
ZONE_INTERVAL = 5000  # Zone appears every 5 seconds
ZONE_DURATION = 3000  # Zone stays for 3 seconds

# Setup screen
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Dodge the Bullets!")

# Font setup
font = pygame.font.SysFont(None, 36)
score = 0

# Player setup
player = pygame.Rect(WIDTH//2, HEIGHT//2, PLAYER_SIZE, PLAYER_SIZE)

# Bullet list
bullets = []
clock = pygame.time.Clock()
running = True
last_difficulty_increase = pygame.time.get_ticks()
start_time = pygame.time.get_ticks()
last_zone_spawn = pygame.time.get_ticks()
zone_active = False
zone_start_time = 0

ZONE_WIDTH, ZONE_HEIGHT = 150, 150
zone = pygame.Rect(0, 0, ZONE_WIDTH, ZONE_HEIGHT)

def spawn_bullet():
    global BULLET_SPEED
    side = random.choice(['top', 'bottom', 'left', 'right'])
    if side == 'top':
        x, y = random.randint(0, WIDTH), 0
        dx, dy = 0, BULLET_SPEED + random.random()
    elif side == 'bottom':
        x, y = random.randint(0, WIDTH), HEIGHT
        dx, dy = 0, -BULLET_SPEED - random.random()
    elif side == 'left':
        x, y = 0, random.randint(0, HEIGHT)
        dx, dy = BULLET_SPEED + random.random(), 0
    else:  # right
        x, y = WIDTH, random.randint(0, HEIGHT)
        dx, dy = -BULLET_SPEED - random.random(), 0
    bullets.append({'rect': pygame.Rect(x, y, BULLET_SIZE, BULLET_SIZE), 'dx': dx, 'dy': dy})

# Game loop
while running:
    screen.fill(WHITE)
    
# Update score
    score_multiplier = 2 if zone_active and player.colliderect(zone) else 1
    score += score_multiplier * (clock.get_time() // 10)  # Increase based on time elapsed
    
    # Handle events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Move player
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT] and player.x > 0:
        player.x -= PLAYER_SPEED
    if keys[pygame.K_RIGHT] and player.x < WIDTH - PLAYER_SIZE:
        player.x += PLAYER_SPEED
    if keys[pygame.K_UP] and player.y > 0:
        player.y -= PLAYER_SPEED
    if keys[pygame.K_DOWN] and player.y < HEIGHT - PLAYER_SIZE:
        player.y += PLAYER_SPEED
    
    # Increase difficulty over time
    current_time = pygame.time.get_ticks()
    if current_time - last_difficulty_increase > DIFFICULTY_INTERVAL:
        BULLET_SPEED += BULLET_INCREMENT
        BULLET_SPAWN_CHANCE += SPAWN_CHANCE_INCREMENT
        last_difficulty_increase = current_time
    
    # Spawn bullets over time
    if random.random() < BULLET_SPAWN_CHANCE:
        spawn_bullet()

    # Handle zone appearance
    if not zone_active and current_time - last_zone_spawn > ZONE_INTERVAL:
        zone.x = random.randint(0, WIDTH - ZONE_WIDTH)
        zone.y = random.randint(0, HEIGHT - ZONE_HEIGHT)
        zone_active = True
        zone_start_time = current_time

    # Remove zone after duration
    if zone_active and current_time - zone_start_time > ZONE_DURATION:
        zone_active = False
        last_zone_spawn = current_time
    
    # Move bullets
    for bullet in bullets[:]:
        bullet['rect'].x += bullet['dx']
        bullet['rect'].y += bullet['dy']
        
        # Check collision
        if player.colliderect(bullet['rect']):
            running = False
        
        # Remove bullets out of bounds
        if (bullet['rect'].x < -BULLET_SIZE or bullet['rect'].x > WIDTH or 
            bullet['rect'].y < -BULLET_SIZE or bullet['rect'].y > HEIGHT):
            bullets.remove(bullet)

    # Draw double score zone if active
    if zone_active:
        zone_surface = pygame.Surface((ZONE_WIDTH, ZONE_HEIGHT), pygame.SRCALPHA)
        zone_surface.fill(BLUE)
        screen.blit(zone_surface, (zone.x, zone.y))
        text = font.render("x2!", True, BLACK)
        screen.blit(text, (zone.x + ZONE_WIDTH // 2 - text.get_width() // 2, zone.y + ZONE_HEIGHT // 2 - text.get_height() // 2))
    
    
    # Draw player
    pygame.draw.rect(screen, BLACK, player)
    
    # Draw bullets
    for bullet in bullets:
        pygame.draw.rect(screen, RED, bullet['rect'])
    
    # Render score
    score_text = font.render(f"Score: {score}", True, BLACK)
    screen.blit(score_text, (10, 10))
    
    # Refresh screen
    pygame.display.flip()
    clock.tick(60)

pygame.quit()
print(f"You scored {score} points!")
time.sleep(2)
exit()
```
