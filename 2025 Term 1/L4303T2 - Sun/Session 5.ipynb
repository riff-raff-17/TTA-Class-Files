```python
def polygon(n, length):
    """"n is number of sides
    length is the length of each side
    """
    angle = 360 / n
    for i in range(n):
        forward(length)
        left(angle)

make_turtle()
polygon(n=100, length=1)
```

```python
import math

#area of a circle = pi * r^2
#circumference = 2pi *r  OR pi * d

def circle(radius):
    circumference = 2 * math.pi * radius
    n = 30 #how precise our circle is
    length = circumference / n
    polygon(n, length)

make_turtle(delay = 0.02)
circle(30)
```

```python
def polyline(n, length, angle):
    for i in range(n):
        forward(length)
        left(angle)

def arc(radius, angle):
    arc_length = 2 * math.pi * radius * angle / 360
    n = 30
    length = arc_length / n
    step_angle = angle / n
    polyline(n, length, step_angle)

def circle(radius):
    arc(radius, 360)

make_turtle(delay = 0.02)
arc(30, 360)
```

```python
from jupyturtle import make_turtle, forward, left, set_color, hide
import math

def spiral_shell(turns, initial_radius, growth_factor, angle_step):
    """Draws a spiral by increasing the radius gradually."""
    radius = initial_radius
    for _ in range(turns):
        arc(radius, angle_step)  # Draw a small arc segment
        radius += growth_factor  # Increase radius to create the spiral effect

def snail():
    make_turtle(delay=0.001)
    hide()
    
    # Shell spiral (more realistic)
    set_color("brown")
    spiral_shell(turns=35, initial_radius=5, growth_factor=1, angle_step=20)
    
    # Snail's body
    set_color("green")
    arc(radius=70, angle=70)  # Body curve
    
    # Snail's head
    circle(radius=10)  # Small circle for the head

# Run the snail function
snail()
```

```python
def triangle(length):
    for i in range(3):
        forward(length)
        left(120)

def draw_pie(n, length):
    angle = 360 / n
    for i in range(n):
        triangle(length)
        left(angle)

make_turtle(delay = 0.02)
draw_pie(6, 50)
```

```python
def polyline(n, length, angle):
    for i in range(n):
        forward(length)
        left(angle)

def arc(radius, angle):
    arc_length = 2 * math.pi * radius * angle / 360
    n = 30
    length = arc_length / n
    step_angle = angle / n
    polyline(n, length, step_angle)

def circle(radius):
    arc(radius, 360)

def petal(radius, angle):
    for i in range(2):
        arc(radius, angle)
        left(180 - angle)

make_turtle(delay = 0.02)
petal(50, 90)
```

```python
def flower(n, radius, angle):
    for i in range(n):
        petal(radius, angle)
        left(360 / n)

make_turtle(delay=0.0002, animate=True)
hide()
flower(23, 50, 90)
draw()
```

```python
colors = 'blue lightblue black skyblue navy indigo'.split()

def color_petal(radius, angle, color):
    set_color(color)
    for i in range(2):
        arc(radius, angle)
        left(180 - angle)

def color_flower(n, radius, angle):
    for i in range(n):
        color_petal(radius, angle, colors[i % len(colors)]) #cycle through colors
        left(360 / n)
        
make_turtle(delay=0.0002)
color_flower(30, 50, 60)
```

```python
def snowflake_branch(size, levels):
    if levels == 0:
        forward(size)
        return

    size /= 3.0
    snowflake_branch(size, levels - 1)
    left(60)
    snowflake_branch(size, levels - 1)
    right(120)
    snowflake_branch(size, levels - 1)
    left(60)
    snowflake_branch(size, levels - 1)

def draw_snowflake(size, levels, sides):
    for i in range(sides):
        set_color(colors[i % len(colors)])
        snowflake_branch(size, levels)
        right(360 / sides)
        
make_turtle(delay=0.02, width=400, height=400)
draw_snowflake(100,3,6)

```

```python
import random

colors = 'blue lightblue black skyblue navy indigo'.split()

def random_painting(steps, step_length):
    for i in range(steps):
        set_color(random.choice(colors))
        forward(step_length)
        turn_angle = random.choice([0, 90, 180, 270])
        left(turn_angle)

make_turtle(delay=0.02, width=300, height=300)    
random_painting(1000, 10)
```

```python
colors = 'blue lightblue black skyblue navy indigo'.split()

def random_painting(steps, step_length):
    penup = True
    for i in range(steps):
        if penup == True:
            pen_down()
            penup=False
        else:
            pen_up()
            penup=True
        set_color(random.choice(colors))
        forward(step_length)
        #turn_angle = random.choice([0, 90, 180, 270])
        turn_angle = random.randint(-180, 180)
        left(turn_angle)

make_turtle(delay=0.02, width=300, height=300)    
random_painting(1000, 10)
```

```python
def draw_ring(color, x, y):
    pen_up()
    forward(x)
    left(90)
    forward(y)
    right(90)
    pen_down()

    set_color(color)
    circle(30)

    #reset position
    pen_up()
    right(90)
    forward(y)
    left(90)
    forward(-x)
    pen_down()

def olympic_logo():
    #top rings
    draw_ring('blue', -60, 30)
    draw_ring('black', 0, 30)
    draw_ring('red', 60, 30)

    #bottom rings
    draw_ring('yellow', -30, 0)
    draw_ring('green', 30, 0)

make_turtle(delay=0.01, height=200)
olympic_logo()
```
