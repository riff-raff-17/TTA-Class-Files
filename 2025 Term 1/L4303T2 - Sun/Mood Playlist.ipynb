```python
import random
from tkinter import Tk, Label, Text, Button, StringVar, END
from textblob import TextBlob

# Sample playlists for different moods
playlists = {
    "positive": [
        "Happy - Pharrell Williams",
        "Can't Stop the Feeling! - Justin Timberlake",
        "Uptown Funk - Mark Ronson ft. Bruno Mars",
        "Good Vibes - Chris Janson"
    ],
    "neutral": [
        "Let It Be - The Beatles",
        "Boulevard of Broken Dreams - Green Day",
        "Yellow - Coldplay",
        "Wake Me Up - Avicii"
    ],
    "negative": [
        "Someone Like You - Adele",
        "Fix You - Coldplay",
        "Hurt - Johnny Cash",
        "Mad World - Gary Jules"
    ]
}

def analyze_sentiment(text):
    """Analyze sentiment and return mood (positive, neutral, or negative)."""
    analysis = TextBlob(text)
    polarity = analysis.sentiment.polarity
    if polarity > 0.2:
        return "positive"
    elif polarity < -0.2:
        return "negative"
    else:
        return "neutral"

def generate_playlist(mood):
    """Generate a playlist based on the mood."""
    return random.sample(playlists[mood], k=len(playlists[mood]))

def get_playlist():
    user_text = input_text.get("1.0", END).strip()
    if not user_text:
        result_var.set("Please enter some text!")
        return
    
    mood = analyze_sentiment(user_text)
    playlist = generate_playlist(mood)
    
    result_var.set(f"Detected mood: {mood.capitalize()}\n\nPlaylist:\n" + "\n".join(playlist))

# Create the main window
root = Tk()
root.title("Mood-Based Playlist Generator")

# Input Text
Label(root, text="Enter your text:").pack(pady=5)
input_text = Text(root, height=5, width=40)
input_text.pack(pady=5)

# Generate Button
Button(root, text="Generate Playlist", command=get_playlist).pack(pady=10)

# Result Label
result_var = StringVar()
result_label = Label(root, textvariable=result_var, wraplength=300, justify="left")
result_label.pack(pady=10)

# Run the GUI loop
root.mainloop()
```
