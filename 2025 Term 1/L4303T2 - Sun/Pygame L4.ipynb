```python
import pygame
import random
import time

# Initialize pygame
pygame.init()

# Constants
WIDTH, HEIGHT = 600, 600
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)
BLUE = (0, 0, 255, 100)  # Transparent blue for double score zone
PLAYER_SIZE = 30
BULLET_SIZE = 10
PLAYER_SPEED = 5
BULLET_SPEED = 3
BULLET_INCREMENT = 0.1
SPAWN_CHANCE_INCREMENT = 0.005
DIFFICULTY_INTERVAL = 3000
BULLET_SPAWN_CHANCE = 0.02
ZONE_INTERVAL = 5000
ZONE_DURATION = 3000

# Setup screen
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Dodge the Bullets!")

# Font setup
font = pygame.font.SysFont(None, 36)
big_font = pygame.font.SysFont(None, 72)

# Player setup
player = pygame.Rect(WIDTH//2, HEIGHT//2, PLAYER_SIZE, PLAYER_SIZE)
bullets = []

# Zone setup
ZONE_WIDTH, ZONE_HEIGHT = 150, 150
zone = pygame.Rect(0, 0, ZONE_WIDTH, ZONE_HEIGHT)

# Game state
score = 0
clock = pygame.time.Clock()
running = False
zone_active = False
zone_start_time = 0
last_difficulty_increase = pygame.time.get_ticks()
last_zone_spawn = pygame.time.get_ticks()

# =========================
# TITLE SCREEN FUNCTION
# =========================
def show_title_screen():
    screen.fill(WHITE)
    title_text = big_font.render("Dodge the Bullets!", True, BLACK)
    instruction_text = font.render("Press SPACE to Start", True, BLACK)
    
    screen.blit(title_text, (WIDTH // 2 - title_text.get_width() // 2, HEIGHT // 2 - 50))
    screen.blit(instruction_text, (WIDTH // 2 - instruction_text.get_width() // 2, HEIGHT // 2 + 20))
    
    pygame.display.flip()

    waiting = True
    while waiting:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                exit()
            if event.type == pygame.KEYDOWN and event.key == pygame.K_SPACE:
                waiting = False

# =========================
# SPAWN BULLET FUNCTION
# =========================
def spawn_bullet():
    side = random.choice(['top', 'bottom', 'left', 'right'])
    if side == 'top':
        x, y = random.randint(0, WIDTH), 0
        dx, dy = 0, BULLET_SPEED + random.random()
    elif side == 'bottom':
        x, y = random.randint(0, WIDTH), HEIGHT
        dx, dy = 0, -BULLET_SPEED - random.random()
    elif side == 'left':
        x, y = 0, random.randint(0, HEIGHT)
        dx, dy = BULLET_SPEED + random.random(), 0
    else:  # right
        x, y = WIDTH, random.randint(0, HEIGHT)
        dx, dy = -BULLET_SPEED - random.random(), 0
    bullets.append({'rect': pygame.Rect(x, y, BULLET_SIZE, BULLET_SIZE), 'dx': dx, 'dy': dy})

# =========================
# MAIN GAME FUNCTION
# =========================
def run_game():
    global running, score, BULLET_SPEED, BULLET_SPAWN_CHANCE, zone_active, zone_start_time, last_zone_spawn, last_difficulty_increase

    score = 0
    bullets.clear()
    running = True

    while running:
        screen.fill(WHITE)

        # Handle events
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

        # Move player
        keys = pygame.key.get_pressed()
        if keys[pygame.K_LEFT] and player.x > 0:
            player.x -= PLAYER_SPEED
        if keys[pygame.K_RIGHT] and player.x < WIDTH - PLAYER_SIZE:
            player.x += PLAYER_SPEED
        if keys[pygame.K_UP] and player.y > 0:
            player.y -= PLAYER_SPEED
        if keys[pygame.K_DOWN] and player.y < HEIGHT - PLAYER_SIZE:
            player.y += PLAYER_SPEED

        # Increase difficulty over time
        current_time = pygame.time.get_ticks()
        if current_time - last_difficulty_increase > DIFFICULTY_INTERVAL:
            BULLET_SPEED += BULLET_INCREMENT
            BULLET_SPAWN_CHANCE += SPAWN_CHANCE_INCREMENT
            last_difficulty_increase = current_time

        # Spawn bullets
        if random.random() < BULLET_SPAWN_CHANCE:
            spawn_bullet()

        # Handle x2 zone
        if not zone_active and current_time - last_zone_spawn > ZONE_INTERVAL:
            zone.x = random.randint(0, WIDTH - ZONE_WIDTH)
            zone.y = random.randint(0, HEIGHT - ZONE_HEIGHT)
            zone_active = True
            zone_start_time = current_time

        if zone_active and current_time - zone_start_time > ZONE_DURATION:
            zone_active = False
            last_zone_spawn = current_time

        # Update score
        score_multiplier = 2 if zone_active and player.colliderect(zone) else 1
        score += score_multiplier * (clock.get_time() // 10)

        # Move and handle bullets
        for bullet in bullets[:]:
            bullet['rect'].x += bullet['dx']
            bullet['rect'].y += bullet['dy']

            if player.colliderect(bullet['rect']):
                running = False

            if (bullet['rect'].x < -BULLET_SIZE or bullet['rect'].x > WIDTH or
                bullet['rect'].y < -BULLET_SIZE or bullet['rect'].y > HEIGHT):
                bullets.remove(bullet)

        # Draw x2 zone
        if zone_active:
            zone_surface = pygame.Surface((ZONE_WIDTH, ZONE_HEIGHT), pygame.SRCALPHA)
            zone_surface.fill(BLUE)
            screen.blit(zone_surface, (zone.x, zone.y))
            text = font.render("x2!", True, BLACK)
            screen.blit(text, (zone.x + ZONE_WIDTH // 2 - text.get_width() // 2, zone.y + ZONE_HEIGHT // 2 - text.get_height() // 2))

        # Draw player
        pygame.draw.rect(screen, BLACK, player)

        # Draw bullets
        for bullet in bullets:
            pygame.draw.rect(screen, RED, bullet['rect'])

        # Draw score
        score_text = font.render(f"Score: {score}", True, BLACK)
        screen.blit(score_text, (10, 10))

        # Refresh screen
        pygame.display.flip()
        clock.tick(60)

# =========================
# START GAME FLOW
# =========================
show_title_screen()
run_game()

# =========================
# GAME OVER SCREEN
# =========================
pygame.quit()
print(f"You scored {score} points!")
time.sleep(2)
exit()
```

```python
import pygame
import random
import time

# Initialize pygame
pygame.init()

# Constants
WIDTH, HEIGHT = 600, 600
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)
BLUE = (0, 0, 255, 100)
PLAYER_SIZE = 30
BULLET_SIZE = 10
PLAYER_SPEED = 0.1
BULLET_SPEED = 3
BULLET_INCREMENT = 0.1
SPAWN_CHANCE_INCREMENT = 0.005
DIFFICULTY_INTERVAL = 3000
BULLET_SPAWN_CHANCE = 0.02
ZONE_INTERVAL = 5000
ZONE_DURATION = 3000

# Setup screen
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Dodge the Bullets!")

# Font setup
font = pygame.font.SysFont(None, 36)
big_font = pygame.font.SysFont(None, 72)

# Player setup
player = pygame.Rect(WIDTH//2, HEIGHT//2, PLAYER_SIZE, PLAYER_SIZE)
bullets = []

# Zone setup
ZONE_WIDTH, ZONE_HEIGHT = 150, 150
zone = pygame.Rect(0, 0, ZONE_WIDTH, ZONE_HEIGHT)

# Game state
score = 0
clock = pygame.time.Clock()
running = False
zone_active = False
zone_start_time = 0
last_difficulty_increase = pygame.time.get_ticks()
last_zone_spawn = pygame.time.get_ticks()

# =========================
# TITLE SCREEN FUNCTION
# =========================
def show_title_screen():
    screen.fill(WHITE)
    title_text = big_font.render("Dodge the Bullets!", True, BLACK)
    instruction_text = font.render("Press SPACE to Start", True, BLACK)
    
    screen.blit(title_text, (WIDTH // 2 - title_text.get_width() // 2, HEIGHT // 2 - 50))
    screen.blit(instruction_text, (WIDTH // 2 - instruction_text.get_width() // 2, HEIGHT // 2 + 20))
    
    pygame.display.flip()

    waiting = True
    while waiting:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                exit()
            if event.type == pygame.KEYDOWN and event.key == pygame.K_SPACE:
                waiting = False

# =========================
# GAME OVER SCREEN FUNCTION
# =========================
def show_game_over_screen():
    screen.fill(WHITE)
    game_over_text = big_font.render("Game Over", True, RED)
    score_text = font.render(f"Score: {score}", True, BLACK)
    restart_text = font.render("Press R to Restart or Q to Quit", True, BLACK)
    
    screen.blit(game_over_text, (WIDTH // 2 - game_over_text.get_width() // 2, HEIGHT // 2 - 60))
    screen.blit(score_text, (WIDTH // 2 - score_text.get_width() // 2, HEIGHT // 2 - 20))
    screen.blit(restart_text, (WIDTH // 2 - restart_text.get_width() // 2, HEIGHT // 2 + 40))
    
    pygame.display.flip()
    
    waiting = True
    while waiting:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                exit()
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_r:
                    waiting = False
                    reset_game()
                if event.key == pygame.K_q:
                    pygame.quit()
                    exit()

# =========================
# RESET GAME FUNCTION
# =========================
def reset_game():
    global score, bullets, BULLET_SPEED, BULLET_SPAWN_CHANCE, zone_active, last_difficulty_increase, last_zone_spawn
    score = 0
    bullets.clear()
    BULLET_SPEED = 3
    BULLET_SPAWN_CHANCE = 0.02
    zone_active = False
    last_difficulty_increase = pygame.time.get_ticks()
    last_zone_spawn = pygame.time.get_ticks()
    player.x, player.y = WIDTH // 2, HEIGHT // 2
    run_game()

# =========================
# SPAWN BULLET FUNCTION
# =========================
def spawn_bullet():
    side = random.choice(['top', 'bottom', 'left', 'right'])
    if side == 'top':
        x, y = random.randint(0, WIDTH), 0
        dx, dy = 0, BULLET_SPEED + random.random()
    elif side == 'bottom':
        x, y = random.randint(0, WIDTH), HEIGHT
        dx, dy = 0, -BULLET_SPEED - random.random()
    elif side == 'left':
        x, y = 0, random.randint(0, HEIGHT)
        dx, dy = BULLET_SPEED + random.random(), 0
    else:
        x, y = WIDTH, random.randint(0, HEIGHT)
        dx, dy = -BULLET_SPEED - random.random(), 0
    bullets.append({'rect': pygame.Rect(x, y, BULLET_SIZE, BULLET_SIZE), 'dx': dx, 'dy': dy})

# =========================
# MAIN GAME FUNCTION
# =========================
def run_game():
    global running, score, BULLET_SPEED, BULLET_SPAWN_CHANCE, zone_active, zone_start_time, last_zone_spawn, last_difficulty_increase

    running = True

    while running:
        screen.fill(WHITE)

        # Handle events
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

        # Mouse-controlled movement
        mouse_x, mouse_y = pygame.mouse.get_pos()
        player.x += (mouse_x - player.x - PLAYER_SIZE // 2) * PLAYER_SPEED
        player.y += (mouse_y - player.y - PLAYER_SIZE // 2) * PLAYER_SPEED

        # Keep player in bounds
        player.x = max(0, min(WIDTH - PLAYER_SIZE, player.x))
        player.y = max(0, min(HEIGHT - PLAYER_SIZE, player.y))

        # Increase difficulty over time
        current_time = pygame.time.get_ticks()
        if current_time - last_difficulty_increase > DIFFICULTY_INTERVAL:
            BULLET_SPEED += BULLET_INCREMENT
            BULLET_SPAWN_CHANCE += SPAWN_CHANCE_INCREMENT
            last_difficulty_increase = current_time

        # Spawn bullets
        if random.random() < BULLET_SPAWN_CHANCE:
            spawn_bullet()

        # Handle zone logic
        if not zone_active and current_time - last_zone_spawn > ZONE_INTERVAL:
            zone.x = random.randint(0, WIDTH - ZONE_WIDTH)
            zone.y = random.randint(0, HEIGHT - ZONE_HEIGHT)
            zone_active = True
            zone_start_time = current_time

        if zone_active and current_time - zone_start_time > ZONE_DURATION:
            zone_active = False
            last_zone_spawn = current_time

        # Update score
        score_multiplier = 2 if zone_active and player.colliderect(zone) else 1
        score += score_multiplier * (clock.get_time() // 10)

        # Move bullets
        for bullet in bullets[:]:
            bullet['rect'].x += bullet['dx']
            bullet['rect'].y += bullet['dy']
            if player.colliderect(bullet['rect']):
                running = False
                show_game_over_screen()
            if (bullet['rect'].x < -BULLET_SIZE or bullet['rect'].x > WIDTH or 
                bullet['rect'].y < -BULLET_SIZE or bullet['rect'].y > HEIGHT):
                bullets.remove(bullet)

        # Draw player and zone
        pygame.draw.rect(screen, BLACK, player)
        for bullet in bullets:
            pygame.draw.rect(screen, RED, bullet['rect'])
        if zone_active:
            pygame.draw.rect(screen, BLUE, zone)

        # Display score
        score_text = font.render(f"Score: {score}", True, BLACK)
        screen.blit(score_text, (10, 10))

        pygame.display.flip()
        clock.tick(60)

# =========================
# START GAME FLOW
# =========================
show_title_screen()
run_game()
pygame.quit()
```
