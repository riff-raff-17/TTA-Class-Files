```python
from ugot import ugot
got = ugot.UGOT()
got.initialize("192.168.0.2")
print("Connected!")
```

```python
import time
from IPython.display import clear_output

# Load our color recognition model
got.load_models(["color_recognition"])

got.balance_start_balancing()

# Get color info
try: 
    while True:
        color = got.get_color_total_info()
        if color[0]:
            print(color[0])
        else:
            print("No color!")

        if color[0] == "Red":
            # Forwards is 0, Backwards is 1
            got.balance_move_speed(direction=0, speed=30)
        elif color[0] == "Purple":
            # Left is 2, Right is 3
            got.balance_turn_speed(turn=2, speed=45)
        else:
            got.balance_stop_balancing()

        # set a delay
        time.sleep(0.5)
        clear_output(wait=True)

except KeyboardInterrupt:
    print("Done")
    got.balance_stop_balancing()
```
