```python
from jupyturtle import *
import math
```

### Introduction to jupyturtle

- Canvas is 300x150 pixels by default
- Turtle starts at (150, 75)

```python
make_turtle()
forward(50)
left(90)
back(50)
right(90)
```

```python
make_turtle()
set_width(10)
for i in range(4):
    forward(50)
    left(90)

make_turtle(delay=0)
set_width(10)
for i in range(3):
    forward(50)
    left(120)
```

```python
def polygon(n, length):
    angle = 360 / n
    for i in range(n):
        forward(length)
        left(angle)

make_turtle(width=500, height=500, delay=0.02)
polygon(n=5, length=100)
```

```python
def circle(radius):
    circumference = math.pi * 2 * radius
    n = 30 # precision
    length = circumference / n
    polygon(n, length)

make_turtle(delay=0.02)
circle(radius=30)
```

```python
make_turtle(delay=0.0001)
for i in range(24):
    circle(30)
    left(15)
```

```python
from ugot import ugot
got = ugot.UGOT()
got.initialize("192.168.1.188")
print("Connected!")
```

```python
# Strafe movement
# angle: -180 to 180, speed: speed: 5 to 80, times: 0 - 360, unit: 0 is seconds, 1 is cm
got.mecanum_translate_speed_times(angle=0, speed=40, times=50, unit=1)

# Turning movement
# turn: 2 is left, 3 is right, speed: 5 to 280, times: 0 to 360, unit: 0 is seconds, 2 is degrees
got.mecanum_turn_speed_times(turn=2, speed=45, times=90, unit=2)
```
