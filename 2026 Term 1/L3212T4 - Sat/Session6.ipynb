```python
# Read the words in front of the robot, and print them out to the console
from ugot import ugot
from IPython.display import clear_output
got = ugot.UGOT()
got.initialize('192.168.1.89')

got.load_models(["word_recognition"])
```

```python
try:
    while True:
        text = got.get_words_result()
        print(text)
        clear_output(wait=True)

        if "Forward" in text:
            # 0 is forwards, 1 is back
            got.mecanum_move_speed(0, 30)
        elif "Left" in text:
            # 2 is left, 3 is right
            got.mecanum_turn_speed(2, 45)
        elif "Right" in text:
            got.mecanum_turn_speed(3, 45)
        else:
            got.mecanum_stop()
except KeyboardInterrupt:
    got.mecanum_stop()
    print("Done!")
```

```python
# Commands using a dictionary and lambda functions
# Lambda functions are small, anonymous functions
command_dict = {
    "Forward" : lambda : got.mecanum_move_speed(0, 30),
    "Backward" : lambda : got.mecanum_move_speed(1, 30),
    "Left" : lambda : got.mecanum_turn_speed(2, 45),
    "Right" : lambda : got.mecanum_turn_speed(3, 45),
}

try:
    while True:
        text = got.get_words_result()
        if text in command_dict:
            command_dict[text]
        else:
            got.mecanum_stop()
except KeyboardInterrupt:
    got.mecanum_stop()
```
