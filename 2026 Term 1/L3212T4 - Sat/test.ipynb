```python
from ugot import ugot
from IPython.display import clear_output
got = ugot.UGOT()
got.initialize('192.168.1.1')
```

```python
import time

got.load_models(['face_attribute', 'face_recognition'])

try:
    while True:
        # [gender, mask_info, emotion, center_x, center_y, height, width, area]
        face_char = got.get_face_characteristic_total_info()

        if face_char:
            print("Gender:", face_char[0][0])
            print("Mask:", face_char[0][1])
            print("Emotion:", face_char[0][2])
            
            emotion = face_char[0][2]
            if emotion == "Happy":
                got.mecanum_move_speed(0, 30)
            elif emotion == "Calm":
                got.mecanum_turn_speed(0, 45)
            else:
                got.mecanum_stop()

        else:
            print("No face ._.")
            got.mecanum_stop()

        time.sleep(0.2)
        clear_output(wait=True)

except KeyboardInterrupt:
    print("Done")
    got.mecanum_stop()
```
