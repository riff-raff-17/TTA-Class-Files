## 2. Data Types and Variables

Common data types:
- `int` (whole numbers): `5`, `-2`
- `float` (decimals): `3.14`, `-0.5`
- `str` (text): `"hello"`
- `bool` (True/False): `True`, `False`

Variables store values:
```python
age = 16
name = "Sam"
```

You can check the type with `type`:

```python
type(age) = int
type(name) = string
```

````python
age = input("Enter your age: ")
print("Age:", age)
print("Type of age:", type(age))
````

````python
milk = int(input("How much is milk?"))
eggs = int(input("How much is eggs?"))

milk + eggs
````

## 3. Operators

### Arithmetic
- `+` add
- `-` subtract
- `*` multiply
- `/` divide (always float)
- `//` integer division
- `%` remainder (mod)
- `**` exponent

### Comparison (results in bool)
- `==`, `!=`, `<`, `<=`, `>`, `>=`

### Logical
- `and`, `or`, `not`


````python
print("7 / 2 = ", 7 / 2)
print("7 // 2 = ", 7 // 2)
print("7 % 2 = ", 7 % 2)
````

## 4. if / elif / else

Conditional logic lets your program make decisions.

```python
if condition:
    ...
elif another_condition:
    ...
else:
    ...
```


````python
temp = 10

if temp < 0:
    print("Freezing!")
elif temp < 20:
    print("Cool")
elif temp < 30:
    print("Warm")
else:
    print("Hot!")
````

## 5. Loops

Loops repeat code.

### `while` loop
Repeats *while* a condition is True.

### `for` loop
Repeats over a sequence (often `range(...)` or a list).


````python
count = 1
while count <= 5:
    print("Count:", count)
    count += 1
````

````python
for i in range(1, 6):
    print("i =", i)
````

````python
for i in range(1, 30):
    if i % 2 == 0:
        print(i, "Even")
    else:
        print(i, "Odd")
````

## 6. Functions

Functions help you:
- reuse code
- organize logic
- make programs easier to read and test

```python
def name(parameters):
    # body
    return value  # optional


````python
# DEFINE a function
def greet(name):
    return f"Hello, {name}!"

# RUN a function
message = greet("Jolie")
print(message)
````

````python
def rect_area(width, height):
    area = width * height
    return area

area = rect_area(7, 8)
print(area)
````
