```python
from ugot import ugot
from IPython.display import clear_output
import time
import random

got = ugot.UGOT()
got.initialize("192.168.1.232")
```

```python
# In the UGOT robot, it is called "Stone", "Scissor", "Cloth"

# 'options' is a list of valid moves in Rock-Paper-Scissors
options = ["stone", "scissor", "cloth"]

while True:
    robot_choice = random.choice(options)
    # strip and lower remove spaces and lowercase!
    player_choice = input("What do YOU choose?").strip().lower() 
    
    if player_choice == "quit":
        break
    if player_choice not in options:
        print("Invalid move!")
    elif player_choice == "stone" and robot_choice == "scissor" or \
        player_choice == "scissor" and robot_choice == "cloth" or \
        player_choice == "cloth" and robot_choice == "stone":
        print("You win!")

    elif player_choice == robot_choice:
        print("Draw!")

    else:
        print("You lose!")
        break
```

```python
got.load_models(['gesture'])

try:
    while True:
        options = ["stone", "ok", "cloth"]

        robot_choice = random.choice(options)
        # strip and lower remove spaces and lowercase!
        while True:
            player_choice = got.get_gesture_result().strip().lower()
            if player_choice in options:
                break

        text = f"I played {robot_choice} and you played {player_choice}"
        got.play_audio_tts(text, voice_type=0, wait=True)

        if player_choice == "stone" and robot_choice == "scissor" or \
            player_choice == "scissor" and robot_choice == "cloth" or \
            player_choice == "cloth" and robot_choice == "stone":
            print("You win!")
            got.play_audio_tts("You win!", voice_type=0, wait=True)

        elif player_choice == robot_choice:
            print("Draw!")
            got.play_audio_tts("Draw!", voice_type=0, wait=True)

        else:
            print("You lose!")
            got.play_audio_tts("You lose!", voice_type=0, wait=True)

except KeyboardInterrupt:
    got.balance_stop_balancing()
    print("Done!")
```
