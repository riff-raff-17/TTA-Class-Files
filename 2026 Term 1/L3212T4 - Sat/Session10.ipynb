```python
# Open your camera

import cv2
from IPython.display import display, clear_output
from PIL import Image
import numpy as np
from ugot import ugot

# Make one instance of the UGOT class
got = ugot.UGOT()
got.initialize("192.168.1.1")

got.open_camera()

def display_frame():
    frame = got.read_camera_data()
    if frame is not None: # Checks if the frame is real!
        nparr = np.frombuffer(frame, np.uint8) # Reads in the image data
        data = cv2.imdecode(nparr, cv2.IMREAD_COLOR) # Turns numbers into color
        frame_rgb = cv2.cvtColor(data, cv2.COLOR_BGR2RGB)

        display(Image.fromarray(frame_rgb))
        clear_output(wait=True)
```

```python
try:
    while True:
        display_frame()

except KeyboardInterrupt:
    print("Done!")
```

```python
# In the UGOT Robot it is called "Stone", "Scissor", "Cloth"
import random

# "options" is a list of valid moves in RPS
options = ['stone', 'scissor', 'cloth']

while True:
    robot_choice = random.choice(options)
    # strip and lower remove spaces and lowercase!
    player_choice = input("What do YOU choose?").strip().lower()

    if player_choice == "quit":
        break
    if player_choice not in options:
        print("Invalid move!")
    elif player_choice == "stone" and robot_choice == "scissor" or \
    player_choice == "scissor" and robot_choice == "cloth" or \
    player_choice == "cloth" and robot_choice == "stone":
        print("You win!")

    elif player_choice == robot_choice:
        print("Draw!")

    else:
        print("You lose!")
    
```

```python
got.load_models(["gesture"])

try:
    while True:
        options = ["stone", "ok", "cloth"]

        robot_choice = random.choice(options)

        while True:
            player_choice = got.get_gesture_result().strip().lower()
            if player_choice in options:
                break

        text = f"I played {robot_choice} and you played {player_choice}"
        got.play_audio_tts(text, voice_type=0, wait=True)

except KeyboardInterrupt:
    got.mecanum_stop()
    print("Doneeeee")
```
