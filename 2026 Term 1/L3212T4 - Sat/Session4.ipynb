```python
import cv2
from IPython.display import display, clear_output
from PIL import Image
import numpy as np
from ugot import ugot

got = ugot.UGOT()
got.initialize('192.168.1.204')

got.open_camera()

# A frame is one screenshot
# tinyurl.com/coding2026
def display_frame():
    frame = got.read_camera_data()
    if frame is not None:
        nparr = np.frombuffer(frame, np.uint8)
        data = cv2.imdecode(nparr,cv2.IMREAD_COLOR)
        frame_rgb = cv2.cvtColor(data, cv2.COLOR_BGR2RGB)

        display(Image.fromarray(frame_rgb))
        clear_output(wait=True)

```

```python
try:
    while True:
        display_frame()

except KeyboardInterrupt:
    print("Done")
```

```python
# How to display a frame
# tinyurl.com/coding2026
import time

got.load_models(["color_recognition"])

try:
    while True:
        color = got.get_color_total_info()
        if color[0]:
            print(color[0])
        else:
            print("No color!")

        if color[0] == "Purple":
            got.screen_display_background(2) # 2 is purple
            got.balance_move_speed(direction=0, speed=30) # 0: Forwards, 1: Backwards
        elif color[0] == "Red":
            got.screen_display_background(3) # 3 is red
            got.balance_turn_speed(turn=2, speed=45) # 2: Left, 3: Right
        elif color[0] == "Yellow":
            got.screen_display_background(5) # 5 is yellow
            got.balance_turn_speed(turn=3, speed=45)
        else:
            got.screen_display_background(0) # 0 is black

        display_frame()
except KeyboardInterrupt:
    print("Done")
```
