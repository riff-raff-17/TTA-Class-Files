## 1.1 What is a Variable?

A **variable** is a name that refers to a value stored in the computer's memory. You create one using the `=` sign:

```python
name = value
```

**Important:** In programming, `=` does NOT mean "equals" like in math.  
It means **"store this value into this variable"**.  
Read it as *"gets"* or *"becomes"*.

So `x = 5` means: *"x gets the value 5"*.

````python
# Creating a variable and printing it
x = 5
print(x)
````

````python
# Variables can hold different types of values
age = 14  # integer (whole number)
height = 1.65  # float (decimal number)
name = "Alex"  # string (text — must be in quotes)
is_student = True  # boolean (True or False)

print(age)
print(height)
print(name)
print(is_student)
````

### Variables can be reassigned

A variable's value can change at any time. When you reassign, the **old value is gone forever**.

````python
score = 10
print("Before:", score)

score = 25  # the old value (10) is now replaced
print("After:", score)
````

### The key pattern: using a variable's own value to update itself

This is the pattern that appears on **every** HCIC test. Look at this carefully:

```python
x = 5
x = x + 3
```

This does NOT mean "x equals x plus 3" (that would be nonsense in math!).

It means:
1. **Right side first:** Take the current value of `x` (which is 5), add 3 → that gives 8
2. **Then assign:** Store 8 back into `x`

So after this line, `x` is now **8**.

````python
x = 5
print("x starts at:", x)

x = x + 3     # take x (5), add 3, store result (8) back into x
print("After x = x + 3:", x)

x = x * 2     # take x (8), multiply by 2, store result (16) back into x
print("After x = x * 2:", x)
````

## 1.2 Arithmetic Operators

Python supports these arithmetic operators:

| Operator | Name | Example | Result |
|----------|------|---------|--------|
| `+` | Addition | `7 + 3` | `10` |
| `-` | Subtraction | `7 - 3` | `4` |
| `*` | Multiplication | `7 * 3` | `21` |
| `/` | Division | `7 / 3` | `2.333...` |
| `//` | Floor division | `7 // 3` | `2` |
| `%` | Modulo (remainder) | `7 % 3` | `1` |
| `**` | Exponent (power) | `7 ** 3` | `343` |

Let's look at each one.

````python
# The basics: +, -, *
print("7 + 3 =", 7 + 3)
print("7 - 3 =", 7 - 3)
print("7 * 3 =", 7 * 3)
````

### Division: `/` vs `//`

This is a common source of mistakes on the test.

- `/` **always gives a float** result (it will always have a decimal point)
- `//` gives the **floor** (rounds down to nearest whole number)

**HCIC test trap:** `70 / 5` gives `14.0` (a float!), NOT `14`. This was tested in HCIC 2022.

````python
# Regular division ALWAYS gives a float
print("7 / 3 =", 7 / 3)  # 2.333...
print("70 / 5 =", 70 / 5)  # 14.0  <-- still a float!
print("10 / 2 =", 10 / 2)  # 5.0   <-- still a float!

print()

# Floor division rounds down to an integer
print("7 // 3 =", 7 // 3)  # 2  (drops the decimal part)
print("70 // 5 =", 70 // 5)  # 14 (integer this time)
print("10 // 2 =", 10 // 2)  # 5  (integer)
````

### Modulo: `%` (the Remainder Operator)

The `%` operator gives you the **remainder** after division.  
This is extremely useful and appears on the test often.

Think of it this way:
- `7 % 3` → 7 ÷ 3 = 2 remainder **1** → answer is `1`
- `10 % 3` → 10 ÷ 3 = 3 remainder **1** → answer is `1`
- `12 % 4` → 12 ÷ 4 = 3 remainder **0** → answer is `0`

**Common use on the test:** checking if a number is even or odd:
- `num % 2 == 0` → number is **even**
- `num % 2 == 1` → number is **odd**

````python
print("7 % 3 =", 7 % 3)  # remainder of 7 / 3 = 1
print("10 % 3 =", 10 % 3)  # remainder of 10 / 3 = 1
print("12 % 4 =", 12 % 4)  # remainder of 12 / 4 = 0 (divides evenly)
print("15 % 5 =", 15 % 5)  # remainder of 15 / 5 = 0

print()

# Even or odd?
print("10 % 2 =", 10 % 2, "→ even!")  # 0 means even
print("11 % 2 =", 11 % 2, "→ odd!")  # 1 means odd
print("51 % 2 =", 51 % 2, "→ odd!")  # 1 means odd
````

### Exponent: `**`

`**` raises a number to a power.

- `2 ** 3` means 2³ = 2 × 2 × 2 = 8

````python
print("2 ** 3 =", 2 ** 3)     # 8
print("5 ** 2 =", 5 ** 2)     # 25
print("10 ** 0 =", 10 ** 0)   # 1 (anything to the power of 0 is 1)
````

## 1.3 Compound Assignment Operators

These are **shorthand** for updating a variable using its own value. They appear constantly on the HCIC test.

| Shorthand | Equivalent to | Meaning |
|-----------|--------------|----------|
| `x += 3` | `x = x + 3` | Add 3 to x |
| `x -= 3` | `x = x - 3` | Subtract 3 from x |
| `x *= 3` | `x = x * 3` | Multiply x by 3 |
| `x /= 3` | `x = x / 3` | Divide x by 3 |

They do **exactly the same thing** as the long form. They're just shorter to write.

````python
# Demonstrating each compound operator
score = 10
print("Start:", score)

score += 5  # same as: score = score + 5
print("After += 5:", score)

score -= 3  # same as: score = score - 3
print("After -= 3:", score)

score *= 2  # same as: score = score * 2
print("After *= 2:", score)
````

### Why this matters for the test

HCIC questions often use `+=` inside loops or in multi-step calculations. Here's a pattern straight from the **HCIC 2021 test (Q11)**:

```python
X = 5
Y = 6
X += Y
print(X)
```

Step through it:
1. `X = 5` → X is 5
2. `Y = 6` → Y is 6
3. `X += Y` → same as `X = X + Y` → `X = 5 + 6` → X is **11**
4. Output: **11** → Answer is **(C)**

````python
# Verify!
X = 5
Y = 6
X += Y
print(X)
````

## 1.4 Multiple Assignment

Python lets you assign values to **multiple variables in one line**. This appears frequently on HCIC tests.

```python
a, b = 10, 20
```

This is the same as writing:
```python
a = 10
b = 20
```

The values on the right are matched to the names on the left, in order.

````python
a, b = 10, 20
print("a =", a)
print("b =", b)
````

````python
# Three variables at once
x, y, z = 1, 2, 3
print(x, y, z)
````

### Variable Swapping — A Critical HCIC Pattern

This is a **very common test pattern**. Look at this carefully:

```python
x, y, z = 5, 10, 15
x, y, z = y, z, x
```

The key rule: **Python evaluates the entire right side first**, using the *old* values, and then assigns all at once.

Step through it:
1. After line 1: `x=5`, `y=10`, `z=15`
2. Line 2 — right side first (using old values): `y` is 10, `z` is 15, `x` is 5
3. So line 2 becomes: `x, y, z = 10, 15, 5`
4. Result: `x=10`, `y=15`, `z=5`

This is exactly how a question from **HCIC 2024** works!

````python
x, y, z = 5, 10, 15
print("Before swap:", x, y, z)

x, y, z = y, z, x
print("After swap: ", x, y, z)
````

### A simple two-variable swap

Swapping two variables is a classic pattern:

```python
a, b = 3, 7
a, b = b, a     # swap!
# now a=7, b=3
```

````python
a, b = 3, 7
print("Before:", a, b)

a, b = b, a
print("After: ", a, b)
````

## 1.5 Order of Operations

Python follows the standard mathematical order of operations:

1. **Parentheses** `()` — done first
2. **Exponent** `**`
3. **Multiplication / Division / Floor division / Modulo** `*`, `/`, `//`, `%` — left to right
4. **Addition / Subtraction** `+`, `-` — left to right

When in doubt, **use parentheses** to make things clear!

````python
# Multiplication before addition
print(2 + 3 * 4)        # 3*4=12, then 2+12 = 14  (NOT 20!)

# Parentheses change the order
print((2 + 3) * 4)      # 2+3=5, then 5*4 = 20

# Left to right for same-priority operators
print(10 - 3 + 2)       # 7 + 2 = 9  (left to right)
print(12 / 3 * 2)       # 4.0 * 2 = 8.0  (left to right)
````

## 📝 Key Takeaways

Before moving on to Part 2, make sure you're confident with all of these:

**1. `=` means assignment, not equality.**  
The right side is evaluated first, then stored into the left side.

**2. `/` always gives a float.**  
Even `10 / 2` gives `5.0`, not `5`. Use `//` if you want an integer.

**3. `%` gives the remainder.**  
`num % 2 == 0` checks if a number is even. This shows up on almost every HCIC test.

**4. Compound operators (`+=`, `-=`, `*=`) are shorthand.**  
`x += 3` is exactly the same as `x = x + 3`.

**5. Multiple assignment evaluates the right side first.**  
`a, b = b, a` swaps the values because Python reads all the old values on the right before assigning.

**6. Trace code line by line.**  
Keep a table of variable values. Update it after every line. This is the skill the test is really testing!
