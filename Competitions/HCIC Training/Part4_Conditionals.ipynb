## 4.1 Basic `if` / `else`

The simplest conditional asks one question and takes one of two paths:

```python
if condition:
    # runs if condition is True
else:
    # runs if condition is False
```

**Indentation matters!** The code inside the `if` or `else` block must be indented (4 spaces or 1 tab). Python uses indentation to know which lines belong to which block.

````python
# Basic if/else
score = 75

if score >= 50:
    print("Pass")
else:
    print("Fail")
````

````python
# Change the score and see what happens
score = 30

if score >= 50:
    print("Pass")
else:
    print("Fail")
````

### `if` without `else`

The `else` part is optional. If you leave it out, the program simply does nothing when the condition is False.

````python
x = 3

if x > 10:
    print("x is big")

print("Done")  # this always prints, regardless of the condition
````

## 4.2 Comparison Operators

These operators compare two values and return `True` or `False`:

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `==` | Equal to | `5 == 5` | `True` |
| `!=` | Not equal to | `5 != 3` | `True` |
| `<` | Less than | `3 < 5` | `True` |
| `>` | Greater than | `5 > 3` | `True` |
| `<=` | Less than or equal to | `5 <= 5` | `True` |
| `>=` | Greater than or equal to | `6 >= 5` | `True` |

⚠️ **Critical mistake to avoid:** `=` is assignment, `==` is comparison.  
`x = 5` stores 5 in x.  
`x == 5` asks "is x equal to 5?"

````python
x = 5

print(x == 5)  # True  — is x equal to 5?
print(x == 6)  # False — is x equal to 6?
print(x != 3)  # True  — is x not equal to 3?
print(x < 10)  # True  — is x less than 10?
print(x >= 5)  # True  — is x greater than or equal to 5?
print(x > 5)  # False — is x strictly greater than 5?
````

### Comparing strings

You can use `==` and `!=` to compare strings too. Remember: **case sensitive**!

- `'Hello' == 'Hello'` → `True`
- `'Hello' == 'hello'` → `False` (capital H vs lowercase h)

````python
colour = "black"

if colour == "black":
    print("It's black!")
else:
    print("Not black")

print()

# Case matters:
print("Hello" == "Hello")  # True
print("Hello" == "hello")  # False
````

## 4.3 `elif` — Multiple Branches

When you have more than two possibilities, use `elif` (short for "else if"):

```python
if condition1:
    # runs if condition1 is True
elif condition2:
    # runs if condition1 is False AND condition2 is True
elif condition3:
    # runs if condition1 and condition2 are False AND condition3 is True
else:
    # runs if ALL conditions above are False
```

**Important:** Python checks the conditions **in order from top to bottom** and runs the **first one that is True** — then skips the rest.

````python
score = 72

if score >= 75:
    print("Distinction")
elif score >= 50:
    print("Pass")
else:
    print("Fail")
````

````python
# Try different scores
for score in [90, 72, 49, 75, 50]:
    if score >= 75:
        result = "Distinction"
    elif score >= 50:
        result = "Pass"
    else:
        result = "Fail"
    print(f"Score {score}: {result}")
````

### ⚠️ Key trap: only the FIRST matching branch runs

Once Python finds a True condition, it runs that block and **skips everything else** in the chain.

````python
# Both conditions are True for x=15, but only the FIRST runs
x = 15

if x > 10:
    print("greater than 10")  # This runs
elif x > 5:  # This is never checked because the first was True
    print("greater than 5")
else:
    print("5 or less")
````

## 4.4 Logical Operators: `and`, `or`, `not`

These let you combine multiple conditions into one.

### `and` — both conditions must be True

| A | B | A and B |
|---|---|--------|
| True | True | **True** |
| True | False | False |
| False | True | False |
| False | False | False |

### `or` — at least one condition must be True

| A | B | A or B |
|---|---|-------|
| True | True | **True** |
| True | False | **True** |
| False | True | **True** |
| False | False | False |

### `not` — flips True to False and vice versa

| A | not A |
|---|------|
| True | False |
| False | True |

````python
# and — both must be True
x = 7
print(x > 5 and x < 10)  # True AND True  → True
print(x > 5 and x > 10)  # True AND False → False
````

````python
# or — at least one must be True
x = 7
print(x > 5 or x > 10)  # True OR False → True
print(x > 8 or x > 10)  # False OR False → False
print(x < 3 or x > 5)  # False OR True → True
````

````python
# not — flips the result
x = 7
print(not x > 5)  # not True → False
print(not x > 10)  # not False → True
````

## 4.5 Truthiness — What Python Considers True or False

In Python, you can use almost any value as a condition — not just `True` and `False`. This is called **truthiness**.

### Values that are **falsy** (treated as False):
- `False`
- `0` (integer zero)
- `0.0` (float zero)
- `''` (empty string)
- `[]` (empty list)
- `None`

### Everything else is **truthy** (treated as True):
- Any non-zero number: `1`, `-5`, `3.14`
- Any non-empty string: `'hello'`, `'0'`, `'False'` ← **note: the string 'False' is truthy!**
- Any non-empty list

This is a common test trap: `'False'` (a non-empty string) is **truthy**, even though it looks like it should be false.

````python
# Falsy values
falsy_values = [False, 0, 0.0, "", [], None]
for v in falsy_values:
    if v:
        print(f"{repr(v)} is TRUTHY")
    else:
        print(f"{repr(v)} is FALSY")
````

````python
# Truthy values — including some surprising ones!
truthy_values = [True, 1, -1, 3.14, "hello", "0", "False", [0]]
for v in truthy_values:
    if v:
        print(f"{repr(v)} is TRUTHY")
    else:
        print(f"{repr(v)} is FALSY")
````

### ⚠️ The BIG trap: `'False'` is truthy!

- `False` (no quotes, boolean) → **falsy** → `if False:` does NOT run
- `'False'` (with quotes, string) → **truthy** → `if 'False':` DOES run

This exact distinction appeared in **HCIC 2022 B3** and **HCIC 2023 B3**.

````python
# Boolean False — falsy
if False:
    print("Boolean False is truthy")  # does NOT print
else:
    print("Boolean False is falsy")  # prints

# String 'False' — truthy!
if "False":
    print("String 'False' is truthy")  # prints!
else:
    print("String 'False' is falsy")
````

## 4.6 The `is` Operator vs `==`

These two are easy to confuse:

- `==` checks if two values are **equal** (same value)
- `is` checks if two variables point to the **exact same object in memory**

For short strings and small integers, Python often reuses the same object in memory, so `is` and `==` can give the same result. But they are fundamentally different.

**The test question:** "If `a = 'False'`, what is `b` for `print(a is b)` to give `True`?"

The answer is `b = 'False'` — b must be assigned the **same string**.

This appeared in **HCIC 2022 B3**.

````python
# == checks value equality
a = [1, 2, 3]
b = [1, 2, 3]
print(a == b)  # True — same values
print(a is b)  # False — different objects in memory!
````

````python
# For strings, Python often reuses objects
a = "False"
b = "False"
print(a == b)  # True
print(a is b)  # True (Python reuses short string objects)
````

````python
# HCIC 2022 B3: what is b for print(a is b) to give True?
a = "False"
b = "False"  # must be the same string
print(a is b)
````

## 4.7 The Modulo Trick for Even/Odd

This is one of the most tested patterns on HCIC. The modulo operator `%` gives the remainder, and checking `num % 2` tells you if a number is even or odd:

- `num % 2 == 0` → **even** (divides evenly by 2)
- `num % 2 == 1` → **odd** (has remainder 1 when divided by 2)

This appeared in **HCIC 2022 B5** and **HCIC 2023 Q14**.

````python
# Basic even/odd check
num = 51

if num % 2 == 0:
    print(num, "is even.")
else:
    print(num, "is odd.")
````

````python
# You can also test with != 0
num = 51

if num % 2 != 0:
    print(num, "is odd.")
else:
    print(num, "is even.")
````

### The modulo trick generalises

- `num % 3 == 0` → divisible by 3
- `num % 5 == 0` → divisible by 5
- `num % n == 0` → divisible by n

And from **HCIC 2023 B2**:
- `item % 3 == 2` → remainder is 2 when divided by 3

````python
# HCIC 2023 B2 pattern
for item in [1, 2, 3, 4, 5, 6]:
    if item % 3 == 2:
        print(item, end=" ")  # which numbers have remainder 2 when divided by 3?
print()
````

## 4.8 Nested Conditionals

You can put an `if` statement **inside** another `if` statement. This is called nesting.

```python
if outer_condition:
    if inner_condition:
        # runs only if BOTH are True
    else:
        # runs if outer is True but inner is False
else:
    # runs if outer is False
```

Be careful with indentation — each level of nesting adds more indentation.

````python
score = 82
bonus = True

if score >= 75:
    if bonus:
        print("Distinction with bonus!")
    else:
        print("Distinction")
elif score >= 50:
    print("Pass")
else:
    print("Fail")
````

## 4.9 Conditionals Inside Loops

The HCIC test frequently combines loops and conditionals. A `for` loop runs the same block of code repeatedly, and inside that loop an `if` statement makes a decision for each item.

This is the same filter-and-collect pattern from Part 3, but now applied to lists.

````python
# Count how many numbers in a list are greater than 5
numbers = [3, 7, 2, 9, 1, 6, 4]
count = 0

for n in numbers:
    if n > 5:
        count += 1

print(count)  # how many were > 5?
````

````python
# Collect items that meet a condition
numbers = [3, 7, 2, 9, 1, 6, 4]
big = []

for n in numbers:
    if n > 5:
        big.append(n)

print(big)
````

## 📝 Key Takeaways

**1. `=` is assignment, `==` is comparison.**  
This is the most common typo in Python. `if x = 5:` is a syntax error. You need `if x == 5:`.

**2. Only the first matching `elif` branch runs.**  
Once Python finds a True condition in an `if/elif/else` chain, it runs that block and skips the rest — even if later conditions would also be True.

**3. `and` requires both to be True. `or` requires at least one.**  
`5 > 3 and 5 > 10` → False (second part is False).  
`5 > 3 or 5 > 10` → True (first part is True).

**4. Falsy values: `0`, `False`, `''`, `[]`, `None`.**  
Everything else is truthy. `'False'` (string) is truthy. `False` (boolean) is falsy.

**5. The modulo pattern for even/odd: `num % 2 == 0`.**  
If the remainder after dividing by 2 is 0, the number is even. This shows up on every HCIC paper.

**6. Fill-in-the-condition questions: work backwards.**  
Look at which inputs produce which outputs. Find what property the "True" inputs have in common. That's your condition.

**7. Case sensitivity matters in string comparisons.**  
`'A' == 'a'` is False. Always check the case of strings in conditions.
