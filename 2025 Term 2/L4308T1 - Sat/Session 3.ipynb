```python
from textblob import TextBlob
from textblob import Word
```

```python
def emo_checker():
    print("Hello, I am a 100% accurate emotion checker!")
    running = True

    while running:
        user_input = input("How are you feeling today? >")
        blob = TextBlob(user_input)
        polarity = blob.sentiment.polarity

        if 'exit' in user_input:
            break
        else:
            if polarity == 1.0:
                print('Extremely happy')
            elif polarity > 0.8:
                print("Super happy")
            elif polarity > 0.6:
                print("Pretty happy") 
            elif polarity > 0.4:
                print("Happy")
            elif polarity > 0.2:
                print("Slightly happy")
            elif polarity > 0:
                print('ok')
            elif polarity == 0:
                print('Neutral')
            elif polarity > -0.2:
                print('slightly sad')
            elif polarity > -0.4:
                print('sadge')
            elif polarity > -0.6:
                print('pretty sad')
            elif polarity > -0.8:
                print('super sad')
            elif polarity > -1:
                print('Madly sad')
            
            print(polarity)

emo_checker()
        
```

```python
from textblob.sentiments import NaiveBayesAnalyzer

def emo_checker():
    print("Hello, I am a 100% accurate emotion checker!")
    running = True

    while running:
        user_input = input("How are you feeling today? >")
        blob = TextBlob(user_input, analyzer=NaiveBayesAnalyzer())
        polarity = blob.sentiment.p_pos

        if 'exit' in user_input:
            break
        else:
            if polarity == 1.0:
                print('Extremely happy')
            elif polarity > 0.8:
                print("Super happy")
            elif polarity > 0.6:
                print("Pretty happy")
            elif polarity > 0.4:
                print("Happy")
            elif polarity > 0.2:
                print("Slightly happy")
            elif polarity > 0:
                print('ok')
            elif polarity == 0:
                print('Neutral')
            
            print(polarity)

emo_checker()
        
```

```python
import random

def choose_word():
    words = ["gundam"]
    return random.choice(words)

def display(word, guessed_letters):
    return " ".join(letter if letter in guessed_letters else "_" for letter in word)

def hangman():
    word = "apricity zachary"
    guessed_letters = set()
    attempts = 10

    print("Welcome to Hangman!")
    while attempts > 0:
        print("\nWord:", display(word, guessed_letters))
        print("Guessed letters:", ", ".join(sorted(guessed_letters)) if guessed_letters else "None")

        guess = input("Guess a letter or the full word: ").lower()

        if guess == word:
            print("\nCongratulations! You guessed the word:", word)
            break
        elif len(guess) == 1 and guess in guessed_letters:
            print("You already guessed that letter.")
        elif len(guess) == 1 and guess in word:
            guessed_letters.add(guess)
            if set(word).issubset(guessed_letters):
                print("\nCongratulations! You guessed the word:", word)
                break
        else:
            guessed_letters.add(guess)
            attempts -= 1
            print("Incorrect! Attempts left:", attempts)

    else:
        print("\nGame over! The word was:", word)

hangman()
```
