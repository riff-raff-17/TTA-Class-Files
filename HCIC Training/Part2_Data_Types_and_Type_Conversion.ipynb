## 2.1 The Four Main Data Types

Every value in Python has a **type**. The four you need for the HCIC test are:

| Type | Name | Examples | Description |
|------|------|---------|-------------|
| `int` | Integer | `5`, `-3`, `0`, `100` | Whole numbers (no decimal point) |
| `float` | Float | `3.14`, `-0.5`, `14.0` | Numbers with a decimal point |
| `str` | String | `'hello'`, `"5"`, `'True'` | Text — always in quotes |
| `bool` | Boolean | `True`, `False` | Only two values: True or False |

The key insight: **quotes make something a string**.
- `5` is an integer
- `'5'` is a string (it just happens to look like a number)
- `True` is a boolean
- `'True'` is a string

````python
# These look similar but are DIFFERENT types
a = 5  # integer
b = 5.0  # float
c = "5"  # string
d = True  # boolean
e = "True"  # string (NOT a boolean!)

print(a, type(a))
print(b, type(b))
print(c, type(c))
print(d, type(d))
print(e, type(e))
````

### The `type()` function

`type()` tells you what kind of value you're dealing with. This appeared on the **HCIC 2023 test (Q12)**.

The output format is `<class 'typename'>`. So:
- `type(5)` → `<class 'int'>`
- `type('hello')` → `<class 'str'>`

````python
print(type(42))
print(type(3.14))
print(type("hello"))
print(type(False))
````

## 2.2 Why Types Matter: `+` and `*` Behave Differently!

This is where students get tricked on the test. The `+` and `*` operators do **different things** depending on the type:

### With numbers (int/float): normal math
```python
5 + 3     →  8
5 * 3     →  15
```

### With strings: concatenation and repetition
```python
'5' + '3'   →  '53'    (glues them together — NOT 8!)
'5' * 3     →  '555'   (repeats 3 times — NOT 15!)
```

This distinction is tested in **every single HCIC paper**.

````python
# With NUMBERS: + means addition
print(5 + 3)  # 8
print(10 + 20)  # 30
````

````python
# With STRINGS: + means concatenation (gluing together)
print("5" + "3")  # '53' — NOT 8!
print("hello" + " " + "world")  # 'hello world'
````

````python
# With NUMBERS: * means multiplication
print(5 * 3)  # 15
````

````python
# With STRINGS: * means repetition
print("5" * 3)  # '555' — NOT 15!
print("ha" * 4)  # 'hahahaha'
print("ab" * 2)  # 'abab'
````

### The combination rule: string * int works, string * string does NOT

- `'hi' * 3` → `'hihihi'` ✅ (string times integer = repetition)
- `'hi' * 'hi'` → **Error** ❌ (you can't multiply two strings)
- `'hi' + 3` → **Error** ❌ (you can't concatenate a string and an integer)

````python
# This works: string * integer
print("hi" * 3)
````

````python
# This causes an ERROR: string + integer
# Uncomment the line below to see the error:

# print('hi' + 3)
````

## 2.3 Type Conversion Functions

Python gives you functions to convert between types:

| Function | What it does | Example |
|----------|-------------|--------|
| `int()` | Converts to integer | `int('5')` → `5` |
| `float()` | Converts to float | `float('3.14')` → `3.14` |
| `str()` | Converts to string | `str(5)` → `'5'` |

These are critical for the HCIC test because many questions mix types and you need to know what the conversion does.

````python
# str() → converts anything to a string
print(str(5))  # '5'
print(str(3.14))  # '3.14'
print(str(True))  # 'True'

print(type(str(5)))  # <class 'str'>
````

````python
# int() → converts to integer
print(int("5"))  # 5 (string → integer)
print(int(3.9))  # 3 (float → integer — TRUNCATES, does NOT round!)
print(int(True))  # 1 (True becomes 1)
print(int(False))  # 0 (False becomes 0)
````

### ⚠️ Common trap: `int()` truncates, it does NOT round

- `int(3.9)` → `3` (cuts off the decimal, doesn't round to 4)
- `int(7.1)` → `7`
- `int(-2.8)` → `-2`

````python
print(int(3.9))  # 3, NOT 4
print(int(7.1))  # 7
print(int(-2.8))  # -2, NOT -3
````

### ⚠️ What you CAN'T convert

Not every conversion works. These will cause errors:

- `int('hello')` → **Error** (can't turn a word into a number)
- `int('3.14')` → **Error** (can't go directly from string-with-decimal to int)
- `float('abc')` → **Error**

This was tested in **HCIC 2023 Q12**: the question asks which lines cause errors.

````python
# These work fine:
print(int("5"))  # 5
print(float("3.14"))  # 3.14
print(str(100))  # '100'
````

## 2.4 The `str()` Trick in Calculations

A very common HCIC pattern is mixing `str()` with arithmetic to produce surprising results. The idea is:

1. Do some **math** first (with numbers)
2. Convert the result to a **string** with `str()`
3. Then use **string concatenation** (`+`) to glue things together

This was tested in **HCIC 2023 B1** and **HCIC 2024 B1**.

````python
# HCIC 2023 B1 pattern
one = 1
zero = 0
result = str(one + zero) + str(zero)
print(result)

# Let's trace it:
# one + zero = 1 + 0 = 1
# str(1) = '1'
# str(zero) = str(0) = '0'
# '1' + '0' = '10' (string concatenation!)
# Output: 10
````

````python
# HCIC 2024 B1 pattern
first = 2
second = 3
output = str(first * second) + str(second - first)
print(output)

# Trace:
# first * second = 2 * 3 = 6
# str(6) = '6'
# second - first = 3 - 2 = 1
# str(1) = '1'
# '6' + '1' = '61'
# Output: 61
````

## 2.5 The Classic HCIC Trap: String Variables That Look Like Numbers

The HCIC test loves giving you code where variables look like numbers but are actually strings. This is probably the single most common trick on the test.

**HCIC 2022 A12** is a perfect example:

````python
# HCIC 2022 A12 — What is the output?
i = "1"  # this is a STRING, not the number 1!
x = "5"  # this is a STRING, not the number 5!
result = 2 * i + x
print(result)
````

Let's trace that step by step:

1. `i = '1'` → i is the **string** `'1'`
2. `x = '5'` → x is the **string** `'5'`
3. `2 * i` → `2 * '1'` → **string repetition** → `'11'`
4. `'11' + x` → `'11' + '5'` → **string concatenation** → `'115'`
5. Output: `115`

If `i` and `x` were integers (no quotes), the answer would have been `2*1 + 5 = 7`. The quotes change everything!

### Let's compare side by side

````python
# WITH quotes — strings
i = "1"
x = "5"
result = 2 * i + x
print("With strings:", result, "  type:", type(result))

# WITHOUT quotes — integers
i = 1
x = 5
result = 2 * i + x
print("With integers:", result, "  type:", type(result))
````

## 2.6 Division Always Produces a Float

This was covered briefly in Part 1, but it's really a **type** issue. In Python 3:

- `/` **always** returns a `float`, even if the result is a whole number
- `//` returns an `int` (if both operands are ints)

This was tested in **HCIC 2022 B4**: "Write down the output for `print(ans)` if `ans = 70/5`."

The answer is `14.0`, not `14`!

````python
ans = 70 / 5
print(ans)  # 14.0
print(type(ans))  # <class 'float'>

print()

ans2 = 70 // 5
print(ans2)  # 14
print(type(ans2))  # <class 'int'>
````

## 2.7 Booleans as Numbers

In Python, booleans are secretly integers:
- `True` = `1`
- `False` = `0`

This means you can do math with them (though this is rarely tested directly).

````python
print(True + True)  # 1 + 1 = 2
print(True + False)  # 1 + 0 = 1
print(True * 5)  # 1 * 5 = 5
print(False * 100)  # 0 * 100 = 0
````

### `'True'` vs `True` — Don't confuse them!

- `True` (no quotes) is a **boolean** — it equals `1` in math
- `'True'` (with quotes) is a **string** — it's just the text "True"

Similarly:
- `False` (no quotes) is a **boolean** — it equals `0` and is falsy
- `'False'` (with quotes) is a **string** — and as a non-empty string, it's actually **truthy**!

This distinction appears in **HCIC 2022 B3** and **HCIC 2023 B3**.

````python
# Boolean True/False
print(type(True))  # <class 'bool'>
print(type(False))  # <class 'bool'>

# String 'True'/'False'
print(type("True"))  # <class 'str'>
print(type("False"))  # <class 'str'>
````

````python
# A key test concept: the 'is' keyword checks identity
# HCIC 2022 B3: if a = 'False', what is b for print(a is b) to give True?

a = "False"
b = "False"  # b must be the same string
print(a is b)  # True — they are the same object
````

## 📝 Key Takeaways

Before moving on to Part 3, make sure you're confident with all of these:

**1. Quotes change everything.**  
`5` is an integer. `'5'` is a string. They behave completely differently with `+` and `*`.

**2. `+` with strings means concatenation.**  
`'5' + '3'` → `'53'`, not `8`. The values are glued together.

**3. `*` with a string and an integer means repetition.**  
`'5' * 3` → `'555'`, not `15`. The string is repeated.

**4. You can't mix types freely.**  
`'hello' + 5` → Error. You need `'hello' + str(5)` or convert the string to a number.

**5. Conversion functions: `int()`, `float()`, `str()`.**  
Use these to switch between types. `int()` truncates floats (doesn't round). `int('hello')` causes an error.

**6. `/` always returns a float.**  
`70 / 5` → `14.0`. This is a type issue — the result is a float even when it divides evenly.

**7. The `str() + str()` pattern.**  
HCIC loves code like `str(a + b) + str(c)` — do the math inside `str()` first, then concatenate the resulting strings.
