## 3.1 Strings Are Sequences of Characters

A string is a sequence of characters. Each character has a **position** (called an index), starting from **0**.

For the string `'Hello'`:

```
Character:   H    e    l    l    o
Index:       0    1    2    3    4
Neg index:  -5   -4   -3   -2   -1
```

Key points:
- Indexing starts at **0**, not 1
- Spaces, punctuation, and digits all count as characters
- You can also count backwards from the end using negative numbers

````python
word = "Hello"
print(len(word))  # 5 characters long

# Each character and its index
for i in range(len(word)):
    print(f"Index {i}: {word[i]}")
````

## 3.2 Indexing: Accessing a Single Character

Use square brackets with an index number to get **one character**:

```python
word[0]    # first character
word[1]    # second character
word[-1]   # last character
word[-2]   # second to last character
```

````python
word = "Python"

# Positive indexing (from the start)
print(word[0])  # P
print(word[1])  # y
print(word[5])  # n (last character — index 5 for a 6-char string)
````

````python
word = "Python"

# Negative indexing (from the end)
print(word[-1])  # n (last character)
print(word[-2])  # o (second to last)
print(word[-6])  # P (same as word[0])
````

### Visualising both index systems

For `'Python'`:

```
Character:    P    y    t    h    o    n
Pos index:    0    1    2    3    4    5
Neg index:   -6   -5   -4   -3   -2   -1
```

Both systems refer to the same characters. Use whichever is more convenient.

## 3.3 Slicing: Extracting a Substring

Slicing lets you extract **a portion** of a string. The syntax is:

```python
word[start:stop]
```

This gives you characters from index `start` **up to but NOT including** index `stop`.

The "up to but not including" rule is the most important thing to remember.

```
word = 'Python'

Character:    P    y    t    h    o    n
Index:        0    1    2    3    4    5

word[0:3] → 'Pyt'   (index 0, 1, 2 — stops BEFORE 3)
word[1:4] → 'yth'   (index 1, 2, 3 — stops BEFORE 4)
word[2:5] → 'tho'   (index 2, 3, 4 — stops BEFORE 5)
```

````python
word = "Python"

print(word[0:3])  # 'Pyt' — characters at index 0, 1, 2
print(word[1:4])  # 'yth' — characters at index 1, 2, 3
print(word[2:5])  # 'tho' — characters at index 2, 3, 4
````

### Omitting start or stop

You can leave out `start` or `stop` to go from the beginning or to the end:

- `word[:3]` → from the start up to index 3 (same as `word[0:3]`)
- `word[3:]` → from index 3 to the end
- `word[:]` → the entire string (a copy)

````python
word = "Python"

print(word[:3])  # 'Pyt' — from start to index 3
print(word[3:])  # 'hon' — from index 3 to end
print(word[:])  # 'Python' — entire string
````

### Slicing with negative indices

Negative indices work in slices too. This is a **very common HCIC pattern**.

```
word = 'This is Hwachong'

word[-9:]  → last 9 characters → ' Hwachong'
word[-4:]  → last 4 characters → 'hong'
```

````python
word = "This is Hwachong"

print(word[-9:])  # ' Hwachong' — last 9 characters (note the space!)
print(word[-4:])  # 'hong'
print(word[:4])  # 'This'
print(word[1:3])  # 'hi'
````

## 3.4 The Step Parameter: `word[start:stop:step]`

Slicing has an optional **third parameter** — the step (or stride). It controls how many positions to skip:

```python
word[start:stop:step]
```

- `word[::2]` → every 2nd character (skip one each time)
- `word[::3]` → every 3rd character
- `word[::-1]` → **reverse** the string (step of -1 goes backwards)

The step parameter appeared in **HCIC 2022 A14** and **HCIC 2023 B5**.

````python
word = "abcdefghijk"

print(word[::2])  # every 2nd character: a, c, e, g, i, k
print(word[::3])  # every 3rd character: a, d, g, j
print(word[1::2])  # starting at index 1, every 2nd: b, d, f, h, j
````

### String Reversal: `[::-1]`

This is a very important pattern. A step of `-1` reverses the string.

This was used in **HCIC 2022 A14** to check for palindromes.

````python
word = "Hello"
print(word[::-1])  # 'olleH' — reversed!

word2 = "racecar"
print(word2[::-1])  # 'racecar' — same! It's a palindrome

# Palindrome check: is a word the same forwards and backwards?
print(word2 == word2[::-1])  # True — racecar is a palindrome
````

## 3.5 The `in` Operator — Membership Testing

The `in` keyword checks whether a character (or substring) exists inside a string. It returns `True` or `False`.

```python
'a' in 'cat'       →  True
'x' in 'cat'       →  False
'at' in 'cat'      →  True (substrings work too)
```

This is used heavily in HCIC questions that iterate through strings and filter characters (HCIC 2022 A13, 2023 Q13, 2024 Q13).

````python
print("a" in "cat")  # True
print("x" in "cat")  # False
print("at" in "cat")  # True — substrings work
print("A" in "cat")  # False — case sensitive!
````

### ⚠️ Case matters!

`'A'` and `'a'` are **different characters**. Python is case-sensitive.

- `'A' in 'Apple'` → `True` (capital A is there)
- `'a' in 'Apple'` → `False` (no lowercase a)

This was specifically tested in **HCIC 2021 Q15** and **HCIC 2024 Q13**.

````python
print("A" in "Apple")  # True
print("a" in "Apple")  # False — no lowercase 'a'!
print("p" in "Apple")  # True
print("P" in "Apple")  # False — no uppercase 'P' after the 'A'
````

## 3.6 Iterating Through Strings with `for`

You can loop through every character in a string using a `for` loop:

```python
for char in 'Hello':
    print(char)
```

This prints: `H`, `e`, `l`, `l`, `o` (one per line).

The most common HCIC pattern combines this with `in` to **filter** characters — checking each character against a set and building a new string.

````python
# Basic iteration
for char in "Hello":
    print(char)
````

### The Classic HCIC Pattern: Filter and Collect

This pattern appears in **HCIC 2021 Q15, 2022 A13, 2023 Q13, and 2024 Q13**. It always works the same way:

1. Start with an empty string (`result = ''`)
2. Loop through each character of a word
3. Check if the character is in some set (like vowels)
4. If yes, add it to the result using `+=`
5. Print the result

````python
# Example: extract all vowels from a word
vowels = "aeiou"
word = "Hwachong"
result = ""

for char in word:
    if char in vowels:
        result += char

print(result)
````

Let's trace that step by step:

```
char = 'H' → 'H' in 'aeiou'? No  → result stays ''
char = 'w' → 'w' in 'aeiou'? No  → result stays ''
char = 'a' → 'a' in 'aeiou'? Yes → result = '' + 'a' = 'a'
char = 'c' → 'c' in 'aeiou'? No  → result stays 'a'
char = 'h' → 'h' in 'aeiou'? No  → result stays 'a'
char = 'o' → 'o' in 'aeiou'? Yes → result = 'a' + 'o' = 'ao'
char = 'n' → 'n' in 'aeiou'? No  → result stays 'ao'
char = 'g' → 'g' in 'aeiou'? No  → result stays 'ao'

Output: ao
```

Notice that `'H'` was NOT collected because `'H'` (uppercase) is not in `'aeiou'` (lowercase). Case matters!

## 3.7 The Palindrome Check Pattern

A palindrome is a word that reads the same forwards and backwards (e.g., 'radar', 'noon', '2002').

In Python, you check this with `word == word[::-1]`.

This was tested in **HCIC 2022 A14**.

````python
# Check some words
words = ["noon", "radar", "2002", "yoyo", "i am,ma i"]

for w in words:
    reversed_w = w[::-1]
    is_palindrome = w == reversed_w
    print(f"'{w}' reversed is '{reversed_w}' → palindrome: {is_palindrome}")
````

## 3.9 Tuple Concatenation with Slicing

Tuples use the same slicing syntax as strings. This appeared in **HCIC 2022 A15**.

A tuple is like a list but uses parentheses `()` and **cannot be changed** (immutable). You can concatenate tuples with `+`.

````python
# Basic tuple slicing
t = ("a", "b", "c", "d", "e")
print(t[1:3])  # ('b', 'c')
print(t[:2])  # ('a', 'b')
print(t[0:])  # entire tuple
````

### Practice Problem 2 — Based on HCIC 2021 Q15

The algorithm counts uppercase 'A' characters across three inputs:

```
Input D
T = 0
for each index in D
    if D[index] == 'A'
        add 1 to T
Output T
```

The three inputs are: `"An Apple A Day"`, `'Make it Happen'`, `"A Friend"`

If you add the 3 outputs together, what is the value?

(A) 7 &emsp; (B) 4 &emsp; (C) 3 &emsp; (D) 2

````python
# Let's trace each one:
def count_A(D):
    T = 0
    for index in range(len(D)):
        if D[index] == "A":
            T += 1
    return T


r1 = count_A("An Apple A Day")
r2 = count_A("Make it Happen")
r3 = count_A("A Friend")

print(f"'An Apple A Day': {r1}")
print(f"'Make it Happen': {r2}")
print(f"'A Friend': {r3}")
print(f"Total: {r1 + r2 + r3}")
````

## 📝 Key Takeaways

**1. Indexing starts at 0.**  
The first character is `word[0]`, not `word[1]`. Negative indices count from the end: `word[-1]` is the last character.

**2. Slicing: `word[start:stop]` — stop is NOT included.**  
`word[1:4]` gives characters at index 1, 2, 3 (not 4). This is the most common mistake.

**3. Omitting start/stop.**  
`word[:3]` = from start. `word[3:]` = to end. `word[:]` = everything.

**4. The step parameter: `word[::n]`.**  
Takes every nth character. `word[::2]` = every other character. `word[::-1]` = reversed.

**5. Case sensitivity matters with `in`.**  
`'A'` and `'a'` are different. `'A' in 'aeiou'` is `False`.

**6. The filter-and-collect pattern.**  
Start with `result = ''`, loop through characters, check with `if char in basket`, add with `result += char`. This appears on every HCIC paper.

**7. Draw out the index table.**  
For any slicing question, write out the characters with their index numbers. This is how you avoid mistakes on the exam.
