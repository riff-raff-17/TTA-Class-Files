```python
from ugot import ugot
import time

got = ugot.UGOT()
got.initialize('192.168.1.106')
got.open_camera()
got.load_models(["line_recognition"])
got.set_track_recognition_line(0)
```

```python
def line_follow(mult=0.25, speed=20):
    offset, line_type, x, y = got.get_single_track_total_info()
    adj_turn = mult * offset
    got.mecanum_move_xyz(x_speed=0, y_speed=speed, z_speed=int(adj_turn))
    return line_type

def pick_up():
    got.mecanum_translate_speed_times(90, 10, 10, 1)
    got.mechanical_clamp_close()
    time.sleep(1)
    got.mechanical_joint_control(0, 45, 45, 1000)
    time.sleep(1)
    got.mechanical_joint_control(0, 0, -65, 1000)
    time.sleep(1)

def knock_away():
    got.mechanical_joint_control(-40, -25, -25, 500)
    time.sleep(1)
    got.mechanical_joint_control(40, -25, -25, 250)
    time.sleep(0.5)
    got.mechanical_joint_control(0, 45, 45, 500)
    time.sleep(1)
```

```python
try:
    got.mechanical_joint_control(0, 45, 45, 500)
    got.mechanical_clamp_close()
    while True:
        line_type = line_follow(mult=0.25, speed=20)
        dist_btm = got.read_distance_data(21)

        if line_type == 0:
            got.mecanum_stop()
            got.mecanum_turn_speed_times(2, 45, 180, 2)
            break

        if dist_btm < 12:
            got.mecanum_stop()
            knock_away()

except KeyboardInterrupt:
    got.mecanum_stop()
    print("Done")
```

```python
'''Just pickup'''

HOLDING = False

try:
    got.mechanical_joint_control(-90, -30, 20, 500)
    got.mechanical_clamp_release()
    time.sleep(1)

    while True:

        line_type = line_follow(mult=0.25, speed=20)
        dist_top = got.read_distance_data(22)

        if line_type == 0:
            got.mecanum_move_speed_times(0, 10, 5, 1)
            got.mecanum_stop()
            got.mechanical_joint_control(0, 0, -10, 500)
            time.sleep(1)
            got.mechanical_clamp_release()
            break

        if dist_top < 15 and not HOLDING:
            got.mecanum_move_speed_times(0, 5, 5, 1)
            got.mecanum_stop()
            pick_up()
            HOLDING = True
            time.sleep(1)
            got.mechanical_joint_control(0, 45, 45, 1000)
            got.mecanum_translate_speed_times(-90, 10, 10, 1)

except KeyboardInterrupt:
    got.mecanum_stop()
    print("Done")
```
